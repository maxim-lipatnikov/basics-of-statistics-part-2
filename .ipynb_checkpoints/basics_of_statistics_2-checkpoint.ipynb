## **1. Анализ номинативных данных**

### **1.1 Постановка задачи**

В прошлом курсе мы фокусировались на **двух основных ситуациях:**
1. У нас есть две или больше **количественные** переменные и мы хотим исследовать **их взаимосвязь** - 
используем **корреляционный или регрессионный анализ**
2. **Сравниваем группы между собой** (по сути, тестируем гипотезу о взаимосвязи номинативной и количественной переменных, 
т.к если, например, между количественной переменной **"рост"** и номинативной переменной **"группа"** есть взаимосвязь - 
это значит, что группы различаются между собой) - используем **t-тест или дисперсионный анализ (ANOVA)**

Но что, если мы хотим исследовать именно **номинативную переменную или ее взаимосвязь с другими номинативными переменными?**
Примеры возможных типов гипотез в анализе номинативных данных:

1. **Гипотеза о распределении номинативной переменной** - хотим проверить, 
есть ли статистически значимые различия между **эмпиричеким** распределением (выборкой) 
и **теоретическим** распределением ГС (чаще всего в качестве теоретического распределения подразумевается **равномерное** - 
когда каждая из градаций номинативной переменной встречается **одинаковое количество раз**)

    **Пример:** есть три партии (А,Б,В), мы спросили 90 человек, за какую **одну** партию они бы проголосовали.
    Единственное, что мы можем сделать - это построить распределение номинативной переменной **"партия"**,
    подсчитав, сколько человек выбрали каждую из партий:

    <img src="images/1.1 hypothesis 1.png" width="300" />

    На графике пунктирной линией обозначен вариант теоретического распределения (все градации равновероятны), 
    и мы четко видим явные отклонения от нее.

    **Задача:** проверить, насколько статистически значимы отклонения от теоретического распределения.

2. **Гипотеза о взаимосвязи двух номинативных переменных** - хотим понять, как градации одной номинативной переменной
влияют на градации другой номинативной переменной в ГС.
    
    **Пример:** взяли 6 пациентов и дали одним лекарство А, другим - лекарство В.
    Вторая номинативная переменная - выздоровел человек (1) или нет (0):
    
    <img src="images/1.1 hypothesis 2.png" width="200" />
    
    Для визуализации такой задачи удобно строить **таблицу сопряженности**, где столбцы - это градации одной переменной,
    а строки - градации другой. В данном случае хочется сделать вывод о том, 
    что лекарство А более действенно, чем В.
    
    **Задача:** проверить, насколько статистически значима взаимосвязь между двумя номинативными переменными
    (в данном примере - насколько значима взаимосвязь в ГС между типом лекарства и тем, выздоровел человек или нет).
    
3. **Гипотеза о взаимосвязи номинативной ЗП и каких-то НП** - хотим оценить взаимосвязь различных НП
(их может быть несколько и они могут быть как номинативными, так и количественными) и номинативной ЗП в ГС.

    Для решения такой задачи, как правило, применяется **логистическая регрессия** (была рассмотрена в прошлой части + 
    будет рассмотрена далее)
    
**Немного о типах переменных:**

Переменная сама по себе не является **ни количественной, ни номинативной и ни ранговой** до тех пор,
пока мы ее не **измерили**.

**Пример:** к какому типу относится переменная "рост"?
- Количественная, если измерили в сантиметрах
- Номинативная, если разделили на группы (например, [130-160] и [160-190])
- Ранговая, если разделили на группы с порядком (например, низкие, средние и высокие)

Как правило, в зависимости от решаемой задачи, бывает удобнее работать не с количественной переменной
(например, **зарплата**), а с номинативной (**доход выше медианы и доход ниже медианы**),
поэтому количественную переменную всегда можно проконвертировать в номинативную.


### **1.2 Расстояние Пирсона**

Итак, рассмотрим проверку гипотезы о соответствии эмпирического распределения номинативной переменной
теоретическому (равномерному) распределению.

Для проверки такой гипотезы необходимо подсчитать **частоту встречаемости градаций номинативной переменной**,
а также разобраться с понятиями:
- **Расстояние** хи-квадрат $\chi^2$ Пирсона
- **Распределение** хи-квадрат $\chi^2$
- **Критерий** (тест) хи-квадрат $\chi^2$ Пирсона

#### 1.2.1 Пример с монеткой
У нас есть монетка, мы подбросили ее **60** раз. Из них **20 раз выпала решка** и **40 раз - орел**.
Мы хотим проверить гипотезу о том, что наша монетка **честная** (т.е. что выпадение орла или решки равновероятны).

**Сформулируем гипотезы:**
- $H_0: p_o = 0.5$ (вероятность выпадения орла равна 0.5 => вероятность выпадения решки тоже 0.5 => исходы **равновероятны**)
- $H_1: p_o \neq 0.5$ (вероятность выпадения орла **не** равна 0.5 => исходы **не равновероятны**)

**Теперь введем обозначения:**
- **Наблюдаемые** в эмпирическом распределении частоты обозначим **O (observed)**
- **Ожидаемые** частоты обозначим **E (expected/estimated)**

Для удобства построим **таблицу сопряженности (contingency table)**,
где обозначим наши наблюдаемые и ожидаемые частоты.
Для расчета ожидаемых значений нужно **взять общее количество наблюдений и поделить на количество исходов/градаций
номинативной переменной**.

В нашем случае, т.к. у монетки только **2 градации**, а у нас всего **60 бросков**,
мы ожидаем увидеть **30 выпавших орлов** и **30 выпавших решек**:

<img src="images/1.2.1 contingency table.png" width="300" />

#### 1.2.2 Общая логика
Логика очень похожа на **t-критерий Стьюдента**, с помощью которого мы проверяем равенство **выборочных средних**.
В t-тесте мы считаем **t-статистику** следующим образом:
берем **разность двух выборочных средних** $\bar{x_1}-\bar{x_2}$ и нормируем ее на **стандартную ошибку $se$**,
в итоге получаем t-статистику.
Полученная **t-статистика** будет иметь **распределение Стьюдента** с df=N-2,
и ее можно использовать для расчета p-value,
т.е. **вероятности получить такие или еще более значимые отклонения случайно при верной Н0**.

**В нашем случае хочется сделать нечто похожее:**
- Как-то обозначить **степень расхождения** между ожидаемыми и наблюдаемыми частотами
- Свести это расхождение к какому-то **одному числу** и сделать выводы о верности $H_0$

Проверим разные способы расчета этой "степени расхождения":

1. **Посчитаем разность между наблюдаемыми и ожидаемыми частотами и сложим их:**

    $(20 - 30) + (40 - 30) = 0$, т.к. у нас отклонения как в большую, так и в меньшую сторону, т.е. нам мешает знак. 
    Значит, такой способ нам не подходит.

2. **Посчитаем разность между наблюдаемыми и ожидаемыми частотами и сложим их квадраты (как в дисперсии):**

    $(20 - 30)^2 + (40 - 30)^2 = 200$. Отлично, мы получили число, которое описывает **степень расхождения**. 
    
    **Но есть подвох:**
    
    Если бы у нас было не 20 решек и 40 орлов, а **1020 решек** и **1040 орлов**, то мы бы тоже получили
    $(1020 - 1030)^2 + (1040 - 1030)^2 = 200$. 
    Получается, что полученное значение может говорить как о **значимых расхождениях** 
    (если у нас 20 раз выпала решка и 40 - орел, то **орел выпадал в два раза чаще**),
    так и о совершенно **незначимых** (как в случае с 1020 решками и 1040 орлами,
    где отклонение на 10 от ожидаемой частоты 1030 не кажется значительным).
    
    **Логика тут такая:** 
    
    Чем **больше** выборка и чем **больше** ожидаемая частота встречаемости градации признака - 
    тем **большие отклонения от ожидаемой частоты можно считать возможными**, 
    поэтому нам необходимо нормировать каждую разность между О и Е на корень из ожидаемой частоты Е:
    
    $(\frac{(20 - 30)}{\sqrt{30}})^2 + (\frac{(40 - 30)}{\sqrt{30}})^2 = \frac{(20 - 30)^2}{30} + \frac{(40 - 30)^2}{30} = \frac{100}{30} + \frac{100}{30} \approx 6.7 $
    
    **Почему нормируем именно на $\sqrt{E}$?**
    
    Пирсон математически доказал, что при условии, что $H_0$ верна,
    разность между $O$ и $E$ пропорциональна $\sqrt{E}$ (подробнее читать [тут](http://bioinformatics.ru/Data-Analysis/Pearson_chisquare_distance.html))

Оказывается, рассчитанное значение - это **расстояние $\chi^2$ Пирсона**:

$$ \chi^2 = \sum_{i=1}^{n} \frac{(O_{i} - E_{i})^2}{E_{i}} $$
, где $O_i$ - observed frequency, $E_i$ - expected frequency

**Если $H_0: p_o = 0.5$ верна**, и мы бы многократно повторяли этот эксперимент и рассчитывали $\chi^2$,
то **он имел бы распределение $\chi^2$ Пирсона**.



### **1.3 Распределение Хи-квадрат Пирсона**

**Попробуем ответить на вопрос:**
если $H_0$ **верна**, и мы будем многократно повторять эксперимент с монеткой (подкинем 60 раз, 
посчитаем кол-во выпавших орлов и решек, потом снова подкинем 60 раз и т.д.) и 
считать **расстояние $\chi^2$**, то как оно будет распределено?

Интуитивно кажется, что распределение будет обладать явно выраженной **асимметрией**, а 
большинство значений сгруппируются **около нуля**, потому что если $H_0$ верна,
то большинство отклонений наблюдаемых частот **О** от ожидаемых частот **Е** будут близко к нулю.

**Проверим с помощью симуляции:**


```python
# Method 1

import random
import matplotlib.pyplot as plt

y = []
for i in range(1000):
    x = []
    for i in range(60):
        x.append(random.randint(0, 1))

    y.append((x.count(0) - 30)**2 / 30 + (x.count(1) - 30)**2 / 30)
    
plt.figure(figsize=(4,3))
plt.hist(y)
plt.show();
```

```python
# Method 2

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chisquare


number_of_experiments = 1000
number_of_throws = 60
expect = [number_of_throws / 2] * 2
chi_result = []
p_result = []

for exp in range(number_of_experiments):
    coin_throw = np.random.randint(0, 2, number_of_throws)
    result = (np.count_nonzero(coin_throw == 0)
              , np.count_nonzero(coin_throw == 1))
    chi, p_value = chisquare(result, f_exp=expect) # calculating chi2 distance
    chi_result.append(chi)
    p_result.append(p_value)

fig, ax = plt.subplots(1, 2, figsize=(8,3))

sns.histplot(chi_result, ax=ax[0], color='red', bins=8, kde=True) # distribution of chi2 distances
ax[0].set_xlim(0, max(chi_result))
ax[0].set_ylabel('Density')

sns.histplot(p_result, ax=ax[1], color='blue', bins=6, kde=True) # distribution of p-values
ax[1].set_xlim(0, 1)
ax[1].set_ylabel('Density');
```

большинство значений **около нуля**. Не трудно догадаться, что это - **распределение Хи-квадрат Пирсона**.

#### 1.3.1 Логика распределения

В нашем случае формула расстояния Пирсона выглядит так:

$$ \chi^2 = \frac{(O_{1} - E_{1})^2}{E_{1}} + \frac{(O_{2} - E_{2})^2}{E_{2}} 
= (\frac{O_{1} - E_{1}}{\sqrt{E_{1}}})^2 + (\frac{O_{2} - E_{2}}{\sqrt{E_{2}}})^2$$

Возьмем первое слагаемое $(\frac{O_{1} - E_{1}}{\sqrt{E_{1}}})^2$ и представим,
что **если $H_0$ верна** (т.е. монетка честная, а значит P(орел) = P(решка) = 0.5).
Тогда, если мы будем **многократно повторять наш эксперимент**, каждый раз рассчитывая,
чему равно $(\frac{O_{1} - E_{1}}{\sqrt{E_{1}}})^2$, то окажется, что $(\frac{O_{1} - E_{1}}{\sqrt{E_{1}}})^2$ 
будет иметь **стандартное нормальное распределение ($\mu$ = 0, $\sigma$ = 1)**.
Аналогичное верно для **каждого слагаемого** (под квадратом).

**Логика:**

Если нулевая гипотеза верна, то логично, что **большинство** значений разности между наблюдаемой и ожидаемой частотами будет
распределено вокруг **нуля**. При этом отклонения у нас могут быть как влево, так и вправо,
потому что разность $O_i - E_i$ может быть как **положительной**, так и **отрицательной**.

Более подробно о том, почему распределение слагаемого под квадратом стандартное нормальное,
можно прочитать [здесь](http://bioinformatics.ru/Data-Analysis/Pearson_chisquare_distance.html).
Вот скриншот из статьи:

<img src="images/1.3.1 chi2 standard normal.jpeg" width="600" />

#### 1.3.2 Определение

**Распределение $ \chi^2 $ (хи-квадрат) с $k$ степенями свободы** — 
это распределение суммы квадратов $k$ **независимых стандартных нормальных случайных величин**.

- **Степени свободы** $k$ - это количество слагаемых, т.е. количество градаций нашей номинативной ЗП
- **Независимость случайных величин** - это когда известное значение одной из них **не дает информации о значении другой**


```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,5)

x = np.linspace(0, 16, 300)
df_set = {1, 2, 3, 4, 7, 10} # degrees of freedom

for df in df_set:
    y = chi2.pdf(x, df=df)
    plt.plot(x, y, label=f'df={df}')

plt.title('Chi-Squared Distribution With Different Degrees of Freedom')
plt.legend()
plt.show();
```

#### 1.3.3 Пример распределения с k=2

Для нагладности визуализируем распределение на плоскости.
Пусть у нас есть две случайных величины - **X** и **Y**.
Они обе **независимы** и имеют **стандартное нормальное распределение ($\mu$ = 0, $\sigma$ = 1)**.
Мы многократно прогоняем эксперимент и получаем множество **пар значений X и Y**, 
которые покажем на плоскости:

<img src="images/1.3.3 example with k=2.png" width="400" />

Рассмотрим точку справа, к которой ведет красная стрелка.
Расстояние от **0** до этой точки - это **гипотенуза** прямоугольного треугольника, где:
- **a** - это проекция расстояния на ось **первой случайной величины**
- **b** - это проекция расстояния на ось **второй случайной величины**

Получается, что **расстояние $\chi^2$ Пирсона** - это **квадрат гипотенузы**, т.е. $a^2 + b^2$.

**Теперь о форме распределения Хи-квадрат:**

Поскольку в рассмотренном примере мы имеем **две стандартных нормальных случайных величины**,
то 95% наблюдений в каждой из них будут сосредоточены в пределах $0 \pm 2\sigma$ - 
на рисунке выше **на плоскости** это показано **синим квадратом**.
Таким образом, если мы для каждой точки будем рассчитывать расстояние Пирсона, 
то чаще всего эти расстояния будут **небольшими**, а соответственно, 
**большинство** "наблюдений" (расстояний Хи-квадрат) будут расположены близко к **нулю**.

Например, 95% наблюдений в распределении Хи-квадрат с 2 степенями свободы лежат в **интервале [0, 5.99)**:

<img src="images/1.3.3 95 percent.png" width="300" />

Таким образом, рассчитав значение $\chi^2$, мы можем либо сравнить его с **критическим значением** распределения $\chi^2$,
либо **посчитать p-value** и сравнить его с заданным уровнем значимости $\alpha$.

#### 1.3.4 Подробнее о степенях свободы

Почему график распределения $\chi^2$ при увеличении **k** становится более симметричным и "сдвигается" вправо?

**Каждая степень свободы k** - это какая-то случайная величина со стандартным нормальным распределением,
т.е. **большинство ее значений близки к нулю**. Квадрат чего-то, **близкого к нулю**, тоже **близок к нулю**.
При этом, в каждой из этих случайных величин могут быть значения, **далекие от нуля** (например, на расстоянии 2 или 3 сигм).
А значит, пусть значения каждой из случайных величин и близки к нулю - 
**чем больше квадратов таких случайных величин мы суммируем - тем дальше мы отклоняемся от нуля**.

*При достаточно **большом** числе степеней свободы **k** распределение Хи-квадрат **стремится к нормальному**!

**Важное уточнение про степени свободы:**

Количество **степеней свободы df** у распределения Хи-квадрат равно количеству стандартных нормальных случайных величин
**минус один** ($df = k-1$). Подробнее об этом дальше.

#### 1.3.5 Связь распределения Хи-квадрат с t и F-распределениями

Вспомним, как рассчитывается значение статистики в тесте Стьюдента (на примере одновыборочного теста) 
и смотрим, как t-статистика связана с $\chi^2$-распределением:

$$ t = \frac{\bar{x} - \mu}{se} 
= \frac{\bar{x} - \mu}{\frac{std}{\sqrt{n}}} 
= \frac{\bar{x} - \mu}{\sqrt{\frac{Var}{n}}} 
= \frac{\bar{x} - \mu}{\sqrt{\sum_{i=1}^n\frac{(\bar{x} - \mu)^2}{n(n-1)}}}
= \frac{\bar{x} - \mu}{\sqrt{\frac{1}{n}\chi^2}} $$

- В знаменателе у нас находится **стандартная ошибка se**, т.е. **выборочное стандартное отклонение std**, 
делённое на **корень из объёма выборки n**. 
- В расчёт стандартного отклонения входит **сумма квадратов отклонений наблюдаемых значений от их среднего значения** - 
то есть **сумма нескольких случайных положительных величин** ($std = \sqrt{Var}$). 
- Из предыдущих степов мы знаем, что **сумма квадратов n случайных положительных величин** может быть описана 
при помощи **распределения хи-квадрат с n-1 степеням свободы**, 
т.к. зная выборочное среднее и n-1 элементов выборки, мы всегда можем точно задать последний элемент! 
- Получается, в **знаменателе t-статистики** у нас спрятано **распределение хи-квадрат**, 
для которого нам и нужно знать количество степеней свободы! 
- Таким образом, **степени свободы в t-распределении на самом деле берутся из распределения хи-квадрат, 
которое спрятано в формуле t-статистики**.

Давайте теперь вспомним формулу F-статистики, которая является главным показателем дисперсионного анализа:

$$ F = \frac{\sum_{j=1}^{k} (\bar{y_{j}} - \bar{y})^2/(k-1)}{\sum_{j=1}^{k}\sum_{i=1}^{j} n(y_{ij} - \bar{y_{j}})^2/(n-k)} $$  

- В числителе находится **сумма квадратов отклонений групповых средних от общего среднего**. 
- В знаменателе у нас **сумма квадратов отклонений наблюдений от внутригрупповых средних**. 
- Таким образом, у нас опять две величины, складывающиеся из **квадратов независимых положительных случайных величин**. 
- Высчитывая F-значение, мы делим одну величину, имеющую распределение **хи-квадрат**, на другую величину, 
имеющую распределение **хи-квадрат**. Каждой из них соответствует **своё количество степеней свободы**. 
- Получается, что распределение F-статистики будет характеризоваться двумя параметрами - 
**степенями свободы для двух распределений хи-квадрат**.


### **1.4 Расчет p-уровня значимости**

#### 1.4.1 Еще подробнее о степенях свободы
    
В примере с монеткой у нас **две градации признака (орел и решка)**. 
Получается, что статистика $\chi^2$ будет иметь распределение хи-квадрат с **2 степенями свободы**?

Нет! Дело в том, что в нашем примере  $(\frac{O_{1} - E_{1}}{\sqrt{E_{1}}})^2$ и $(\frac{O_{2} - E_{2}}{\sqrt{E_{2}}})^2$
**не являются независимыми**. Проверим это на примере:
- **Если выпало 20 решек и 40 орлов, то** 
    - $\frac{O_{1} - E_{1}}{\sqrt{E_{1}}} = \frac{20 - 30}{\sqrt{30}} = \frac{-10}{\sqrt{30}}$
    - $\frac{O_{2} - E_{2}}{\sqrt{E_{2}}} = \frac{40 - 30}{\sqrt{30}} = \frac{10}{\sqrt{30}}$
- **Если выпало 10 решек и 50 орлов, то** 
    - $\frac{O_{1} - E_{1}}{\sqrt{E_{1}}} = \frac{10 - 30}{\sqrt{30}} = \frac{-20}{\sqrt{30}}$
    - $\frac{O_{2} - E_{2}}{\sqrt{E_{2}}} = \frac{50 - 30}{\sqrt{30}} = \frac{20}{\sqrt{30}}$
- **...**

Получается, что **зная одно из слагаемых, мы можем точно узнать второе слагаемое, 
что прямо противоречит определению независимости**.

Другими словами, если мы бросили монетку **1 раз и выпал орел - значит решка в этот бросок уже не выпала**.

Если показать на плоскости все возможные пары значений $\frac{O_{1} - E_{1}}{\sqrt{E_{1}}}$ и $\frac{O_{2} - E_{2}}{\sqrt{E_{2}}}$,
они все окажутся **на одной прямой:**

<img src="images/1.4.1 dependent variables.png" width="400" />

**Тогда сколько же в нашем примере степеней свободы?**

В нашем примере статистика $\chi^2$ имеет распределение хи-квадрат с **одной степенью свободы 
(df = n - 1, где n - число градаций номинативной переменной)**.

Указанная выше логика работает **всегда**, т.е. если мы возьмем игральную кость,
имеющую **6 градаций** (т.е. 6 возможных исходов), то $\chi^2$ будет иметь распределение хи-квадрат с **5 степенями свободы**,
т.к. если у нас всего **60 бросков** и мы знаем, сколько раз **выпало 1, 2, 3, 4 и 5**, 
то мы можем точно вычислить, **сколько раз выпало 6**.

Другими словами, в примере с костью мы имеем **5 независимых** случайных величин, а **шестая определяется этими пятью**.

#### 1.4.2 Считаем p-value и проверяем гипотезу

Итак, мы проверяем гипотезу о том, что **распределение градаций исследуемого признака отклоняется от равномерного**.

Учитывая, что форма распределения зависит от количества степеней свободы, 
**вероятность получить такие или еще большие отклонения от нуля (p-value) тоже зависит от количества степеней свободы**.


```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 10)

p = 0.95
x1 = np.linspace(0, 10, 300)
x2 = np.linspace(0, 20, 300)

fig, axs = plt.subplots(nrows=3, ncols=2)
axs =  axs.flat

for i, df in enumerate(df_set):
    chi_value = chi2.isf(1-p, df=df)
    y = chi2.pdf(x, df=df)
    x = x1 if df <= 4 else x2
    
    axs[i].set_title(f'Chi-Squared with {df} DoF')
    axs[i].plot(x, y, color='k')
    left_x, right_x = x <= chi_value, x >= chi_value
    axs[i].fill_between(x[left_x], y[left_x], color='white')
    axs[i].fill_between(x[right_x], y[right_x], color='green')


fig.suptitle('P-value = 0.05 for Different DoF ')
plt.show();
```

### **1.5 Анализ таблиц сопряженности**

Ранее мы проверяли гипотезу о том, что **эмпирическре распределение соответствует теоретическому**,

Т.е. у нас **две номинативные переменные** (пол и профессия), и каждая из них имеет 2 градации.

Такие данные проще всего представить как раз в виде **таблицы сопряженности**:

<img src="images/1.5.1 table 1.png" width="400" />

**Сформулируем гипотезы:**
- $H_0$: **нет** статистически значимой взаимосвязи между переменными **пол** и **профессия** 
(эмпирическое распределение **не отличается** от теоретического)
- $H_1$: **есть** статистически значимая взаимосвязь между переменными **пол** и **профессия** 
(эмпирическое распределение **отличается** от теоретического)

**Дальше логика такая же, как в случае с одной номинативной переменной:**
- Мы предполагаем, что $H_0$ **верна**, и исходя из этого рассчитываем **ожидаемые** частоты
(по сути, отвечаем на вопрос: "если верна нулевая гипотеза, то какие значения у нас были бы в таблице?")
- Получив **ожидаемые частоты**, рассчитываем расстояние $\chi^2$
(т.е. для каждой ячейки рассчитываем квадрат отклонения **наблюдаемой** частоты от **ожидаемой** и складываем их)
- Полученная статистика $\chi^2$ имеет **распределение хи-квадрат**, 
что позволяет нам посчитать p-value и отклонить/не отклонить $H_0$

#### 1.5.2 Рассчитываем ожидаемые частоты
Для наглядности добавим в нашу таблицу **тоталы** (просуммируем наблюдаемые частоты по столбцам и по строкам):

<img src="images/1.5.2 table 2.png" width="400" />

**Как рассчитать ожидаемые частоты?**

Первая идея - просто взять общее число студентов и поделить на количество "ячеек" в таблице (их у нас 4).
Однако, если мы исходим из **верности** $H_0$ (т.е. взаимосвязи межлу полом и профессией нет), 
это **не значит, что распределение будет равномерным**. Почему?

Потому что если у нас, например, 70% мужчин и 30% женщин, то **отсутствие взаимосвязи между полом и профессией**
будет означать, что **в каждой профессии пропорция мужчин к женщинам 70/30 должна сохраниться**.
Другими словами, отсутствие взаимосвязи означает, что **распределение градаций первого признака
должно быть одинаковым во всех градациях второго признака**.

**Тогда посмотрим, какая доля нашей выборки является мужчинами, а какая женщинами:**
- Мужчин 26 из 41 человека, что примерно равно **63.4%**
- Женщин 15 из 41 человека, что примерно равно **36.6%**

Тогда, если $H_0$ верна, то мы ожидаем увидеть 63.4% мужчин-информатиков и 63.4% мужчин-биологов (аналогично с женщинами):
- Из 24 биологов ожидаем увидеть 15.2 мужчин и 8.8 женщин
- Из 17 информатиков ожидаем увидеть 10.8 мужчин и 6.2 женщин

Эти значения и будут **ожидаемыми (expected, E) частотами**.
Формула для расчета ожидаемых частот выглядит так:

$$ f_{ij} = \frac{f_i \cdot f_j}{N} $$

где $f_i$ и $f_j$ - это общие количества наблюдений по строкам и столбцам.

<img src="images/1.5.2 two tables.png" width="400" />

В коде ожидаемые частоты E можно посчитать так:


```python
from scipy import stats

sex_table = ([15,9], [11,6])
chi2, p_value, df, expected = stats.chi2_contingency(sex_table)
expected
```

### **1.6 Анализ таблиц сопряженности - продолжение**

#### 1.6.1 Расстояние Пирсона и степени свободы

Теперь, рассчитав ожидаемые частоты для каждой из ячеек, 
мы можем **рассчитать расстояние $\chi^2$ по уже известной нам формуле**:

$$ \chi^2 = \sum_{i=1}^{n} \frac{(O_{i} - E_{i})^2}{E_{i}} $$

$$ \chi^2 = \sum_{i=1}^{n} \frac{(15 - 15.2)^2}{15.2} + \frac{(9 - 8.8)^2}{8.8} + \frac{(11 - 10.8)^2}{10.8} + \frac{(6 - 6.2)^2}{6.2} $$

**Сколько степеней свободы имеет распределение полученнной статистики $\chi^2$?**

- Мы знаем **общее количество наблюдений** (т.е. все информатики + все биологи = 41, все юноши + все девушки = 41)
- Мы знаем **общие количества наблюдений по строкам и по столбцам** 
(т.е. знаем, сколько всего информатиков, сколько всего биологов, сколько всего женщин и сколько всего мужчин)
- Мы знаем какую-то **одну ячейку таблицы** (т.е. знаем одно наблюдаемое значение, будь то
количество мужчин-информатиков, женщин-биологов или любое другое пересечение)
- Этих **трех вводных нам достаточно** для того, чтобы **рассчитать все остальные ячейки таблицы**

Таким образом, в таблице сопряженности 2х2 статистика $\chi^2$ имеет **распределение хи-квадрат с 1 степенью свободы**.

**Формула в общем случае:**

$$ df = (n-1)(m-1) $$

где $n$ - количество столбцов (без столбцов с "тоталами"), а $m$ - количество строк (без строк с "тоталами")

#### 1.6.2 Поправка Йейтса

**Поправка Йейтса на непрерывность** – это процедура, 
разработанная для критерия хи-квадрат **при работе с таблицами 2×2**,
и заключающаяся в **вычитании 0.5 из разности между наблюдаемой и ожидаемой частотой**:

$$ \chi^2_{Yates} = \sum_{i=1}^{n} \frac{(|O_{i} - E_{i}| - 0.5)^2}{E_{i}} $$

**Зачем нужна?**

Распределение хи-квадрат **непрерывное**, а данные, используемые в критерии хи-квадрат, **дискретные**, 
и поправка Йейтса используется для того, чтобы исправить это несоответствие.

**Идея поправки**

Идея состоит в том, что **уменьшение значения хи-квадрат приводит к уменьшению вероятности ошибки первого рода 
(ошибочного отвержения нулевой гипотезы)**, т.к. буквально двигает полученное значение статистики ближе к нулю. 

**Мнения о поправке Йейтса**

- Некоторые исследователи считают, что она может приводить к **слишком сильной коррекции с понижением мощности** 
и **повышению вероятности ошибки второго рода (ошибочного неотвержения нулевой гипотезы)**. 
- Некоторые статистики **отвергают поправку Йейтса в принципе**, хотя другие находят ее **полезной в случае разреженных данных, 
особенно если хотя бы в одной ячейке ожидаемая величина меньше 5**.

*Если $ |O_{i} - E_{i}|$ велико, то Поправка не окажет практически никакого влияния,
но чем меньше разность ожидаемой и наблюдаемой частот, тем большее влияние оказывает вычитание 0.5.

*Во многих статистических программах и библиотеках поправка Йейтса применяется автоматически при применении критерия хи-квадрат
в случае с таблицей 2х2.

#### 1.6.3 Считаем p-value и делаем выводы


```python
from scipy import stats

observed_array = [[15,9],[11,6]]

chi2, p_value, df, expected_array = stats.chi2_contingency(observed_array, correction=True) # correction - это поправка Йейтса

print(f'''
Chi-Squared Statistic: {chi2}
p-value: {p_value}
Degrees of Freedom: {df}
Expected values: 
{expected_array}
''')

if p_value > 0.05:
    print(f"p-value = {p_value} > 0.05 => Can't Reject H0")
else:
    print(f"p-value = {p_value} < 0.05 => Reject H0")
```

#### 1.6.4 Итоги по критерию Хи-квадрат Пирсона

**Зачем нужен и какую гипотезу проверяет?**

Критерий $\chi^2$ проверяет **гипотезу о соответствии наблюдаемого (эмпирического)
и ожидаемого (теоретического) распределения градаций одной или нескольких номинативных переменных**.

Критерий может применяться в двух типах задач (и формулировка гипотезы немного меняется):
- Если у нас только **одна номинативная переменная**, 
то мы проверяем гипотезу о соответствии эмпирического распределения **равномерному распределению** 
(когда все градации этой переменной встречаются в ГС с одинаковой частотой).
    
    В данном случае статистика $\chi^2$ имеет $df = n-1$ степеней свободы, где $n$ - это количество градаций переменной.
    
- Если у нас **две номинативные переменные**,
то мы проверяем гипотезу о наличии взаимосвязи между переменными
(формально, мы проверяем ту же гипотезу о соответствии эмпирического расппределения теоретическому,
но в данном случае **теоретическое не является равномерным**)

    В данном случае статистика $\chi^2$ имеет $df = (n-1)(m-1)$ степеней свободы, 
    где $n$ - количество столбцов (градации одной переменной), а $m$ - количество строк (градации второй переменной).

*Для применения критерия Хи-квадрат крайне желательно, чтобы количество **ожидаемых наблюдений**
в каждой из ячеек было **не меньше 5**.

**Интерпретация результатов**

Легко заметить, что полученное значение статистики $\chi^2$ - это **одно** число,
которое не дает нам никакой информации о том, 
**в какой конкретно ячейке разница между наблюдаемой и ожидаемой частотами велика**.
Таким образом, если, например, мы **отклоняем нулевую гипотезу**,
то единственный вывод, который мы можем сделать - **распределение не отличается от теоретического**
или **взаимосвязи между переменными нет**.
Более того, как и в случае с **корреляцией, наличие взаимосвязи не означает наличие причинно-следственной связи!**

Для решения этой проблемы проводят **анализ остатков**.

#### 1.6.5 Анализ остатков для более конкретных результатов

Посмотрим на примере задачи:

<img src="images/1.6.5 task for residuals analysis.png" width="300" />

Мы хотим иметь возможность **наглядно оценить, насколько велика разница между наблюдаемой и ожидаемой частотой
в каждой из ячеек таблицы сопряженности**.

Рассчитаем критерий Хи-квадрат и построим **mosaic plot**:


```python
from statsmodels.graphics.mosaicplot import mosaic
from itertools import product
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.patches as mpatches
import numpy as np

cure = ['Placebo', 'Aspirine']
effect = ['Normal', 'Trombosis']
idx = pd.MultiIndex.from_tuples(product(cure, effect))
data = pd.Series([7, 18, 13, 6], index=idx)

def chi2_mosaic(data):
    # Analysis
    data = data.copy()
    observed = data.unstack().T.values
    chi2, p_value, dof, expected = chi2_contingency(observed)
    residuals = (observed - expected)  / np.sqrt(expected)
    
    print(f'''
    Chi-Squared Statistic: {chi2}
    p-value: {p_value}
    Degrees of Freedom: {df}
    Expected values: 
    {expected}
    ''')

    if p_value > 0.05:
        print(f"p-value = {p_value} > 0.05 => Can't Reject H0")
    else:
        print(f"p-value = {p_value} < 0.05 => Reject H0")
    
    # Coloring
    hex_colors = sns.color_palette("coolwarm", 6).as_hex()
    color_labels = ['<-4', '-4:-2', '-2:0', '0:2', '2:4', '>4'][::-1]
    colors = pd.Series([x  for x in residuals.ravel()], index=data.index, name='color')
    valrange = np.array([-np.inf, -4, -2, 0, 2, 4, np.inf])
    color_prop = {k: {'color' : hex_colors[np.argmax(v < valrange) - 1]}  for k, v in colors.iteritems()}
    patches = [mpatches.Patch(color=c, label=l) for c, l in zip(hex_colors, color_labels)]
    
    # Plotting
    fig, ax = plt.subplots(figsize=(5,4))
    mosaic(data, gap=.02, properties=color_prop, ax=ax, labelizer= lambda x: '')
    fig.legend(handles=patches, bbox_to_anchor=(0.6, 0., 0.5, 0.5))
    plt.show();
    
chi2_mosaic(data)
```

**Логика чтения графика:**
- **Размер** прямоугольников соответствует **количеству наблюдений** 
(всего людей, принимавших плацебо, больше, чем тех, кто принимал аспирин)
- **Цвет** прямоугольников соответствует величине **разницы между наблюдаемой и ожидаемой частотой в ячейке, 
заданной в стандартизированной Z-шкале** (в данном случае если отклонение наблюдаемой частоты от ожидаемой **меньше** нуля - 
цвет красный, а если **больше** нуля - цвет синий. Чем больше отклонение - тем **ярче** цвет)
- Если значения стандартизированных остатков больше 3 - можно считать, что в этой ячейке зафиксированы значимые отклонения.

### **1.7 Точный критерий Фишера**

**Точный критерий Фишера** - это статистический тест, который следует использовать в случае,
когда **данных мало** (т.е. когда количество **ожидаемых** наблюдений в каких-либо ячейках таблицы
сопряженности **меньше пяти**).
Сам критерий проверяет гипотезу о **взаимосвязи переменных**.

**Почему не критерий хи-квадрат?**

Потому что если у нас мало данных, мы не можем утверждать, что рассчитанная статистика $\chi^2$ будет иметь
распределение хи-квадрат.

**Пример:**

<img src="images/1.7 table example.png" width="400" />

Как видим, в нашем исследовании всего **8 человек**, и мы хотим исследовать взаимосвязь между лекарством,
которое принимает человек, и тем, выздоровел он или нет.

**Сформулируем гипотезы:**
- $H_0: p_1 = p_2 = p$ - вероятность выздоровления **одинакова** при использовании разных типов лекарств
(**нет взаимосвязи между типом лекарства и выздоровлением**)
- $H_1: p_1 \neq p_2 \neq p$ - вероятность выздоровления **разная** при использовании разных типов лекарств
(**есть взаимосвязь между типом лекарства и выздоровлением**)

**Критерий Фишера позволяет нам ответить на вопрос:**
если $H_0$ **верна**, и мы дадим 4 людям лекарство №1, а другим 4 людям лекарство №2, 
то какова вероятность получить **такие** результаты (т.е. какова вероятность получить **конкретные значения в таблице**,
например, как на картинке выше - (3,1) для лекарства 1 и (1,3) для лекарства 2)?
Т.к. нас интересуют **такие и еще более выраженные отклонения**, 
нас также интересуют исходы (4,0) для лекарства 1 и (0,4) для лекарства 2.

#### 1.7.1 Как работает критерий с точки зрения математики?

Обозначим наблюдаемые частоты так:

<img src="images/1.7.1 math table example.png" width="400" />

$H_0: p_1 = p_2 = p$

**Введем случайные величины:**
- $X$ - количество позитивных исходов ("поправился") у тех, кто принимал **лекарство 1**
- $Y$ - количество позитивных исходов ("поправился") у тех, кто принимал **лекарство 2**

Введенные случайные величины имеют **биномиальное распределение**:
- $X \sim Binomial(a+b,p)$
- $Y \sim Binomial(c+d,p)$
- $X+Y \sim Binomial(n,p)$

**Логика:** по аналогии с монеткой, где у нас есть **2 исхода** (орел и решка) и какая-то вероятность **выпадения орла p**,
в данной задаче у нас тоже **2 исхода** (поправился и не поправился) и какая-то вероятность **выздоровления p**.
Каждый **"бросок монетки"** (т.е. каждый человек в нашем исследовании) - это **распределение Бернулли**,
а т.к. у нас **n человек в выборке** - значит, мы "подбрасываем монетку" **n раз**, а это уже **биномиальное распределение**.

**Теперь рассчитаем вероятности того, что заданные С.В. примут конкретные значения:**
- $P(X=a) = C_{a+b}^a \cdot p^a \cdot (1-p)^b$
- $P(Y=c) = C_{c+d}^c \cdot p^c \cdot (1-p)^d$
- $P(X+Y=a+c) = C_{n}^{a+c} \cdot p^{a+c} \cdot (1-p)^{b+d}$

**Теперь ответим на вопрос:** какова вероятность получить **конкретные значения таблицы** 
(т.е. ровно $a$ выздоровевших среди тех, кто принимал лекарство 1, и $c$ выздоровевших среди тех,
кто принимал лекарство 2)?

$$ P(X=a|X+Y=a+c) = \frac{P(X=a) \cdot P(Y=c)}{P(X+Y=a+c)} $$

**А теперь подставим формулы выше:**

$$ P(X=a|X+Y=a+c) = \frac{P(X=a) \cdot P(Y=c)}{P(X+Y=a+c)} 
= \frac{C_{a+b}^a \cdot p^a \cdot (1-p)^b \cdot C_{c+d}^c \cdot p^c \cdot (1-p)^d}{C_{n}^{a+c} \cdot p^{a+c} \cdot (1-p)^{b+d}} 
= \frac{C_{a+b}^a \cdot C_{c+d}^c}{C_{n}^{a+c}}
= \frac{C_{4}^3 \cdot C_{4}^1}{C_{8}^{4}} \approx 0.229$$

Мы рассчитали вероятность получить **такие отклонения**, 
а теперь рассчитаем вероятность получить **еще более выраженные отклонения**.
В случае **двухсторонней гипотезы** (о наличии взаимосвязи) нам нужно учесть еще 3 случая:
- (4,0) для лекарства 1 и (0,4) для лекарства 2

$$ \frac{C_{a+b}^a \cdot C_{c+d}^c}{C_{n}^{a+c}}
= \frac{C_{4}^4 \cdot C_{4}^0}{C_{8}^{4}} \approx 0.014$$

- (1,3) для лекарства 1 и (3,1) для лекарства 2

$$ \frac{C_{a+b}^a \cdot C_{c+d}^c}{C_{n}^{a+c}}
= \frac{C_{4}^1 \cdot C_{4}^3}{C_{8}^{4}} \approx 0.229$$

- (0,4) для лекарства 1 и (4,0) для лекарства 2

$$ \frac{C_{a+b}^a \cdot C_{c+d}^c}{C_{n}^{a+c}}
= \frac{C_{4}^0 \cdot C_{4}^4}{C_{8}^{4}} \approx 0.014$$

Или же мы просто можем взять вероятности случаев (3,1),(1,3) и (4,0),(0,4) и умножить на 2:

$$ (0.229 + 0.014) \cdot 2 = 0.49 $$

**Проверим вычисление в коде:**


```python
from scipy import stats

odds_ratio, p_value = stats.fisher_exact([[1, 3], [3, 1]])  # odds_ratio - отношение шансов

if p_value > 0.05:
    print(f"p-value = {p_value} > 0.05 => Can't Reject H0")
else:
    print(f"p-value = {p_value} < 0.05 => Reject H0")
```

## **2. Логистическая регрессия и другие непараметрические методы**

### **2.1 Логистическая регрессия. Постановка задачи**

#### 2.1.1 Проблематика и применение функции связи

Основное ограничение **линейной регрессии**, которое не дало бы нам использовать ее в ситуации, 
когда зависимая переменная - **номинативная с двумя градациями (да/нет, 1/0)** - 
это тот факт, что в таком случае в **левой** части формулы регрессии у нас оказывается **переменная с двумя градациями**, 
а в **правой** части мы можем получить **любое значение из [$-\infty,\infty$]**:

$$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

Поскольку нам нужно, чтобы и справа, и слева была **одинаковая "размерность"**,
первый этап решения проблемы - это перейти от **номинативной переменной с двумя градациями (1 и 0)**
хотя бы к **непрерывной переменной - p (вероятность наступления позитивного исхода "1")**:

$$ p = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

Теперь левая часть принимает значения в [0,1], а правая - в [$-\infty,\infty$].
Следующий этап решения проблемы - **переход от вероятности к шансам.**

**Шансы (odds)** - отношение **вероятности успеха** (y=1) к **вероятности неудачи** (y=0).

**Пример:**
пусть мы подбросили монетку **10 раз**, из них **2 раза** выпал орел и **8 раз** решка.

Тогда вероятность выпадения орла:

$$ p_{heads} = \frac{2}{2+8} = 0.2 $$

А шансы вычисляются так:

$$ odds = \frac{p_{heads}}{p_{tails}} = \frac{p_{heads}}{1 - p_{heads}} = \frac{\frac{2}{2+8}}{1 - \frac{2}{2+8}}
= \frac{0.2}{1 - 0.2} = \frac{0.2}{0.8} = 0.25 $$

При этом, **аналогичное** значение можно получить, 
если просто **разделить количество положительных исходов на количество отрицательных**:

$$ odds = \frac{2}{8} = 0.25 $$

#### 2.1.2 Логика понимания шансов

Логично, что если **odds меньше 1** - это говорит о том, что **числитель меньше знаменателя**.
И наоборот, если **odds больше 1** - **числитель больше знаменателя** 
(т.е. вероятность положительного исхода больше, чем вероятность отрицательного).

<img src="images/2.1.2 odds graph.png" width="400" />

Как видно на графике, если мы, например, увидим "шансы равны 150" -
это будет говорить о том, что **вероятность положительного исхода очень близка к 1**.
Если шансы = 0.0001, то **вероятность положительного исхода очень близка к 0**.

Получается, что мы уже перешли от **вероятности**, принимающей значения в [0,1] к **шансам**,
принимающим значения в [0,$\infty$].

#### 2.1.3 Логит-преобразование

А теперь сделаем последний этап - **перейдем от шансов к натуральному логарифму от шансов**:

$$ \ln{(odds)} = \ln{(\frac{p}{1 - p})} = \ln{(0.25)} = -1.38 $$

Логарифмирование позволяет перейти от шансов, принимающих значения в [0,$\infty$],
к логарифму, принимающему значения в [-$\infty$,$\infty$]
(если **шансы меньше 1** - то **логарифм отрицательный**, а если **шансы больше 1** - то **логарифм положительный**):

<img src="images/2.1.3 p vs odds vs ln(odds).png" width="150" />

<img src="images/2.1.3 ln(odds) graph.png" width="400" />

Преобразование, которое мы совершили - называется **логит-трансформацией**,
т.е. применением **функции связи логит**:

$$ \ln{(\frac{p}{1 - p})} = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

Из полученного уравнения мы легко можем **выразить саму зависимую переменную**:

$$ p = \frac{e^{\beta_0 + \beta_1x_1 + ... + \beta_nx_n}}{1+e^{\beta_0 + \beta_1x_1 + ... + \beta_nx_n}} $$

График зависимости вероятности $p$ от натурального логарифма от шансов $\ln{(odds)}$:


```python
# Creating ln(odds)
log_odds = np.arange(-12, 12, 0.1)

# Calculating p from ln(odds)
p = np.exp(log_odds) / (1 + np.exp(log_odds))

plt.figure(figsize=(4,3))
plt.plot(log_odds, p)
plt.grid()
plt.xlim(-12, 12)
plt.ylim(-0.1, 1.1)
plt.xlabel('ln(odds)', fontsize=12)
plt.ylabel('p', fontsize=12)

plt.show();
```

**Как мы так взяли и применили к левой части уравнения какую-то функцию связи?**

На самом деле, основываемся мы на **уравнении линейной регрессии**, которое можно записывать как так:

$$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

так и вот так:

$$ F(y) = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

поскольку в случае линейной регрессии мы тоже применяем функцию связи, но она **ничего не делает (и называется identity)**.
По сути, **регрессионные модели отличаются друг от друга как раз функциями связи** (подробнее об этом в конспекте 1 части курса
в уроке про GLM).

### **2.2 Модель без предикторов. Intercept only model**

Итак, теперь у нас в обеих частях уравнения стоит количественная переменная, 
принимающая значения в [-$\infty$, $\infty$]:

$$ \text{logit}(p) = \ln{\frac{p}{1-p}} = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n $$

Подбор коэффициентов в логистической регрессии осуществляется не методом наименьших квадратов 
(как в линейной регрессии), но смысл тот же - **нам нужно подобрать такие коэффициенты $\beta_0, \beta_1, \beta_2, ... $,
чтобы они наилучшим образом описывали зависимую переменную**.

**Что такое intercept-only модель?** Это регрессионная модель, 
в которой мы **не рассматриваем** влияние **ни одного предиктора** на зависимую переменную.

Рассмотрим данные о пассажирах Титаника и построим модель **без предикторов**,
т.е. воспользуемся только переменной **Survived** и количеством наблюдений.

**Какую гипотезу проверяем?**
- $H_0: \text{intercept} = 0$
- $H_1: \text{intercept} \neq 0$ (intercept статистически значимо отличается от нуля)


```python
import pandas as pd

df = pd.read_csv('data/titanic.csv')
df = df[df['Age'].notnull()]
df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as sf

# Building a model. We use "1" instead of predictors to build an intercept-only model
logreg_no_predictors = sf.glm(formula='Survived ~ 1', data=df, family=sm.families.Binomial()).fit()
logreg_no_predictors.summary()
```

**Логика здесь такая:**
мы строим модель без предикторов, т.е. уравнение регрессии выглядит так 
(мы хотим найти такое одно число $\text{intercept}/$$\beta_0$, чтобы оно наилучшим образом описывало ЗП):

$$ F(y) = \text{intercept} $$

А т.к. в логистической регрессии мы применяем функцию связи **logit**, получаем такую картину:

$$ \text{logit}(p) = \ln{(\frac{p}{1-p})} = \text{intercept} $$

Таким образом, **intercept** - это натуральный логарифм **шансов** положительного исхода.
Как можно видеть, мы получили **coef** = -0.3799 и p-value < 0.05, а значит - **отвергаем $H_0$**.

**Логика гипотезы:** если нулевая гипотеза верна, то intercept = 0. 
Intercept - это натуральный **логарифм** от шансов, а **логарифм равен нулю только если шансы равны единице**.
Шансы же равны единице только в случае, когда p/(1-p) = 1, т.е. когда вероятности положительного
и отрицательного исходов равны.
Таким образом, мы **можем использовать intercept-only логистическую регрессию как аналог критерия хи-квадрат
для случая с одной переменной** (т.е. мы проверяем гипотезу о том, что эмпирическое распределение частот переменной
в ГС равно равномерному).

**Смысл значений std err, z, P>|z|:**
значения, которые мы видим в таблице, получены с помощью ЦПТ.
Пусть нулевая гипотеза верна, и мы многократно берем выборку из ГС и считаем логарифм шансов (intercept). 
В таком случае значения логарифма шансов **распределятся нормально вокруг нуля
со стандартным отклонением std err**. Получается, что если мы возьмем полученное
значение **intercept и поделим на значение std err**, то получим **значение z-статистики**,
что и позволяет нам определить **вероятность получения такого или еще более значимого отклонения от нуля (p-value)**.

Теперь рассчитаем intercept вручную. Для этого сначала построим распределение ЗП:

А теперь найдем шансы того, что случайный пассажир из выборки **выживет**:


```python
# df.Survived.value_counts()[1] - survived (290)
# df.Survived.value_counts()[0] - didn't survive (424)

odds = df.Survived.value_counts()[1] / df.Survived.value_counts()[0]
odds
```

Мы нашли шансы, но **intercept - это натуральный логарифм от шансов**:


```python
import numpy as np

log_odds = np.log(odds)
log_odds
```

Все верно - получили значение $\text{intercept}$ из таблички с результатами модели!
А теперь посмотрим, **как выразить ЗП (вероятность $p$)**.
Для этого сначала возьмем $\text{intercept}$ в качестве **степени экспоненты** 
(логарифмирование наоборот) и получим **шансы**,
а затем выразим вероятность через шансы:


```python
odds = np.exp(log_odds)
odds
```

```python
p = odds / (1 + odds)
p
```

```python
# Or just do it in one step

p = np.exp(log_odds) / (1 + np.exp(log_odds))
p
```

Итак, как уже было указано выше, мы можем использовать intercept-only логистическую регрессию
в качестве аналога критерия хи-квадрат для случая с одной переменной и проверкой гипотезы об отличии
распределения частот переменной от равномерного.

**Важное преимущество логистической регрессии**

Если критерий хи-квадрат позволяет нам сказать, что распределение частот **значимо отличается/не отличается от равномерного**,
то логистическая регрессия позволяет сделать более точный и глубокий вывод:
- Для пассажиров Титаника (ГС) **шансы выжить равны ~68%** (с таким-то доверительным интервалом), 
а **вероятность выжить статистически значимо меньше вероятности погибнуть**.

#### 2.3.1 Интерпретация результатов

Как видим, несмотря на то, что **пола у нас два (мужчины и женщины)**, в модели есть только **intercept и SexMale**.
Как это работает?

1. Когда данные попадают в модель, модель их **"трансформирует" - номинативные переменные разбиваются на несколько колонок**.
Например, переменная **Sex** разбивается на **две** номинативных переменных с градациями 1 и 0 - **SexMale и SexFemale**.
При этом, **если SexFemale = 1, то SexMale всегда равно 0 (и наоборот)**. Получается, что нам в принципе **не нужна одна из переменных**
(а если бы мы использовали обе, мы бы нарушили соблюдение **мультиколлинеарности**).
В нашем случае **основной (референтной) колонкой стала SexFemale** (просто потому что в алфавитном порядке буква F раньше M).
2. **Рассмотрим коэффициенты intercept и SexMale:**
    - **intercept** - это $ \ln{(odds_{female})} $, т.е. просто **натуральный логарифм от шансов выжить у женщин** (т.к.
    наша модель рассматривает женский пол как "референтный")
    - **SexMale** - это $ \ln{(\frac{odds_{male}}{odds_{female}})} $, т.е. **натуральный логарифм 
    отношения шансов выжить у мужчин к шансам выжить у женщин** (опять-таки, 
    мы за "основу" берем шансы женщин и смотрим, "какую часть" от них составляют шансы мужчин)

**Описанную выше логику легче всего показать на примерах:**
- Пусть мы даем нашей модели **женщину** и хотим предсказать, выживет она или нет.
Раз пассажир имеет **женский пол, значит SexMale = 0**.
В таком случае уравнение принимает вид:

$$ \ln{(\frac{p}{1-p})} = \ln{(odds_{female})} + \ln{(\frac{odds_{male}}{odds_{female}})} \cdot \text{SexMale} 
= \ln{(odds_{female})} + \ln{(\frac{odds_{male}}{odds_{female}})} \cdot 0 
= \ln{(odds_{female})} = \text{intercept} = 1.1243 $$

- Теперь мы даем нашей модели **мужчину** и хотим предсказать, выживет он или нет.
Раз пассажир имеет **мужский пол, значит SexMale = 1**.
В таком случае уравнение принимает вид:

$$ \ln{(\frac{p}{1-p})} = \ln{(odds_{female})} + \ln{(\frac{odds_{male}}{odds_{female}})} \cdot \text{SexMale} 
= \ln{(odds_{female})} + \ln{(\frac{odds_{male}}{odds_{female}})} \cdot 1 
= \ln{(odds_{female})}  + \ln{(odds_{male})} - \ln{(odds_{female})} = \ln{(odds_{male})} $$

**Как видим, логика работает** - 
если в модель попадает **женщина**, модель предсказывает шансы ее выживания значением **intercept**,
а если **мужчина**, то добавляется значение **коэффициента перед SexMale** 
(логарифм отношения шансов мы разложили на разность числителя и знаменателя по свойству логарифма).

**Теперь посмотрим на логику самих полученных коэффициентов**, опираясь на таблицу сопряженности.
- **Женщины:** intercept равен **1.1243** (и p-value < 0.05, т.е. intercept статистически
значимо отличается от нуля), т.е. $ \ln{(odds_{female})} = 1.1243 $, т.е. $ odds_{female} = e^{1.1243} \approx 3.08 $.
Если мы проверим шансы по таблице сопряженности, то получим $ odds_{female} = \frac{p_1}{p_0} = \frac{197}{64} \approx 3.08 $.

- **Мужчины:** intercept равен **1.1243**, т.е. $ \ln{(odds_{female})} = 1.1243 $, 
а $\ln{(\frac{odds_{male}}{odds_{female}})} = -2.4778 $ т.е. $ \frac{odds_{male}}{odds_{female}} = e^{-2.4778} \approx 0.08 $.
Получается, что $ odds_{male} = 0.08 \cdot odds_{female} = 0.08 \cdot 3.08 \approx 0.25 $.
Если мы проверим шансы по таблице сопряженности, то получим $ odds_{male} = \frac{p_1}{p_0} = \frac{93}{360} \approx 0.25 $.

    Другими словами, $ \ln{(odds_{male})} = \text{Intercept} + \ln{(\frac{odds_{male}}{odds_{female}})} = 1.1243 - 2.4778 \approx 1.35 = \ln{(\frac{93}{360})} $

#### 2.3.2 Сравнение моделей

Как понять, насколько значимо добавление той или иной переменной в модель?
Для этого мы можем сравнить "нулевую" (intercep-only) модель с моделью с 1 или несколькими предикторами
с помощью показателя **AIC**. **Чем меньше AIC - тем лучше модель.**
Для сравнения можно, например, провести статистический тест для того, чтобы понять,
насколько значимы различия между AIC'ами двух моделей.


```python
print(logreg_no_predictors.aic)
print(logreg_1_predictor.aic)
```

#### 2.3.3 Общие выводы

Как мы видим, **логистическая регрессия** является своеобразным **аналогом критерия хи-квадрат** 
для случая с 2 номинативными переменными, но она **дает гораздо больше информации!**

Если критерий **хи-квадрат позволил бы нам сказать "между полом пассажира и вероятностью выжить есть взаимосвязь"**,
то **логистическая регрессия** позволяет нам сказать, **какие конкретно шансы (а значит и вероятность) выжить
у женщины, насколько "ухудшаются" шансы выжить, если пассажир мужчина 
(т.к. $ \ln{(\frac{odds_{male}}{odds_{female}})}) = \ln{(odds_{male})} - \ln{(odds_{female})} $),
а значит и какие конкретно шансы выжить у мужчин.** 
Более того, логистическая регрессия позволяет **не только делать выводы о ГС**,
но и **прогнозировать шансы/вероятность** выжить для любого интересующего нас пассажира.

Снова прогоним модель с двумя номинативными предикторами:

```python
logreg_2_predictors = sf.glm(formula='Survived ~ C(Sex) * C(Pclass)', data=df, family=sm.families.Binomial()).fit()
logreg_2_predictors.summary()
```

Снова для наглядности построим таблицы сопряженности для женщин и для мужчин:

```python
pd.crosstab(df[df.Sex == 'female'].Survived, df.Pclass)
```

```python
pd.crosstab(df[df.Sex == 'male'].Survived, df.Pclass)
```

**Рассмотрим коэффициенты и их смысл:**
- **C(Sex)[T.male]:C(Pclass)[T.2]:** - это **разность натурального логарифма отношения шансов выжить для мужчин и женщин во 2 классе 
и натурального логарифма шансов выжить для мужчин и женщин в 1 классе**, 
т.е. $\ln{(\frac{odds_{m2}}{odds_{f2}})} - \ln{(\frac{odds_{m1}}{odds_{f1}})} = \ln{(\frac{odds_{m2}}{odds_{f2}})} - (-3.7301) = -0.4204$.

    Логарифм < 0 (p-value > 0.05) => **взаимосвязь между полом и тем, выжил человек или нет,
    проявляется одинаково (т.к. коэффициент значимо не отличается от нуля) в 1 и во 2 классах**.

- **C(Sex)[T.male]:C(Pclass)[T.3]:** - это **разность натурального логарифма отношения шансов выжить для мужчин и женщин в 3 классе 
и натурального логарифма шансов выжить для мужчин и женщин в 1 классе**, 
т.е. $\ln{(\frac{odds_{m3}}{odds_{f3}})} - \ln{(\frac{odds_{m1}}{odds_{f1}})} = \ln{(\frac{odds_{m3}}{odds_{f3}})} - (-3.7301) = 2.1542$.

    Логарифм > 0 (p-value < 0.05) => **взаимосвязь между полом и тем, выжил человек или нет,
    проявляется по-разному в 1 и в 3 классах**.

**Как выглядит итоговое уравнение регрессии, учитывая полученные коэффициенты?**

$$\ln{(\frac{p}{1-p})} = 3.3081 - 3.7301 \cdot \text{SexMale} - 0.8804 \cdot \text{Pclass2} 
- 3.4653 \cdot \text{Pclass3}  - 0.4204 \cdot \text{SexMale:Pclass2} 
+ 2.1542 \cdot \text{SexMale:Pclass3} = 3.3081 - 3.7301 \cdot \text{SexMale} - 0.8804 \cdot \text{Pclass2} 
- 3.4653 \cdot \text{Pclass3}  - 0.4204 \cdot \text{SexMale} \cdot \text{Pclass2} 
+ 2.1542 \cdot \text{SexMale} \cdot \text{Pclass3} $$

Если, например, мы хотим предсказать вероятность выжить для мужчины в 3 классе,
то в уравнении **SexMale = 1, Pclass2 = 0, Pclass3 = 1, SexMale:Pclass2 = 0, SexMale:Pclass3 = 1**.
Другими словами, 

$$\ln{(\frac{p}{1-p})} = 3.3081 - 3.7301 - 3.4653 + 2.1542 = -1.7331 $$

$$ \frac{p}{1-p} = e^{-1.7331} \approx 0.177 $$

$$ 1.177p = 0.177 \Rightarrow p = 0.15 $$

**Логика интерпретации примерно такая:**
- Изначально любому пассажиру мы даем логарифм шансов выжить, равный **intercept**.
- Если мы видим, что пассажир - мужчина, то уменьшаем шансы выжить, вычитая **SexMale**.
- Потом видим, что пассажир не во втором классе, поэтому пропускаем **Pclass2**,
    но зато отнимаем **Pclass3**, тем самым значительно уменьшаем шансы выжить.
- Потом пропускаем **SexMale:Pclass2**, но "компенсируем" шансы выжить за счет прибавления **SexMale:Pclass3**
(т.к. отношение шансов мужчин к шансам женщин в 3 классе значительно выше, чем такое же отношение в 1 классе).

### **2.6 Когда нужно использовать непараметрические методы и почему?**

**Для начала вспомним "необходимые" условия для применения t-теста:**
- **Независимость наблюдений** (требуется вообще везде в статистике)
- **Гомогенность (равенство) дисперсий** (желательно, но необязательно - 
есть в том числе разновидность t-теста для случая с неравными дисперсиями)
- **Нормальное распределение признака в генеральной совокупности** (необязательно)
- **Объем выборки > 30**

Непараметрические тесты нужно использовать тогда, когда не соблюдаются требования для применения параметрических тестов.
Основные требования, насчет которых идут ожесточенные споры - это требование о нормальности распределения и об объеме выборки.

поэтому изложу **свое понимание проблемы, а не то, что в курсе рассказывал Анатолий**.

#### 2.6.1 Требование нормальности распределения

Насколько я понимаю, суть проблемы условия нормальности распределения основывается на 
**устаревшем способе интерпретации ЦПТ:**
- Раньше предполагалось, что **распределение выборочных средних будет иметь нормальное распределение только в случае,
когда ГС имеет нормальное распределение (или близкое к нему)**.
Отчасти это связано и с выбросами - если **ГС распределена нормально**,
значит, **две выборки**, которые мы сравниваем t-тестом, **тоже распределены нормально**,
а значит - в них **нет выбросов**, которые могли бы повлиять на средние $\bar{x_{1}}$ и $\bar{x_{2}}$
и **сделать их распределения ненормальными**. Если распределения выброчных средних **ненормальны** - 
мы не можем быть уверены, что t-статистика имеет распределение Стьюдента.
- Более современная нестрогая интерпретация ЦПТ гласит, что **выборочное среднее в любом случае распределено нормально 
(при достаточном объеме выборки), вне зависимости от того, какое распределение имеет ГС**.

Для t-теста при этом важно, чтобы **среднее (или средние, если сравниваем 2 выборки) имело нормальное распределение**,
но в **устаревшей** формулировке ЦПТ это возможно **ТОЛЬКО если ГС имеет нормальное распределение**.
В современной же формулировке ЦПТ **выборочное среднее ВСЕГДА распределено нормально (или почти нормально)
при достаточном размере выборок**.

В целом, **чем ближе распределение среднего к нормальному** - с тем **большей вероятностью** мы можем быть уверены,
что **t-статистика действительно имеет распределение Стьюдента** (а значит, и с тем большей вероятностью
можем быть уверены в "точности" t-теста).

T-статистика для одновыборочного теста, когда мы сравниваем выборочное среднее с заданной константой (среднее ГС):

$$ t = \frac{\bar{x}-M}{se} = \frac{\bar{x}-M}{\frac{std}{\sqrt{n}}} $$

T-статистика для двухвыборочного теста, когда мы сравниваем 2 выборочных средних:

$$ t = \frac{\bar{x_{1}}-\bar{x_{2}}}{se} = \frac{\bar{x_{1}}-\bar{x_{2}}}{\sqrt{\frac{std_1^2}{n_1} + \frac{std_2^2}{n_2}}} $$

**Итого:**
- **В идеале ГС должна иметь нормальное распределение**, но это совершенно **НЕОБЯЗАТЕЛЬНО** 
(просто в этом случае нас особо **не волнует размер выборки**, т.к. выборочное среднее
при любом размере выборок будет распределено нормально).
Для проверки того, что ГС имеет нормальное распределение (т.е.
проверки гипотезы о том, что наша выборка получена из нормального распределения),
можно использовать **тест Шапиро-Уилка** (о нем дальше) или **тест Колмогорова-Смирнова**.
- **Если мы не знаем распределение ГС (всегда)**, то нам нужно **подобрать такой размер выборки,
чтобы выборочное среднее было распределено нормально (или почти нормально)**,
после чего спокойно проводить t-тест.

#### 2.6.2 "Достаточный" объем выборки

Проблема объема выборки тесно связана с требованием нормальности распределения ГС.
Поскольку мы сделали вывод о том, что для t-теста важно, чтобы **распределение выборочных средних было нормальным**,
нам нужно **подобрать такой размер выборки, при котором распределение среднего наиболее близко к нормальному**.

Для проверки гипотезы о том, что размер выборки должен быть > 30,
проведем эксперимент и посмотрим на распределение выборочных средних, 
если **мы берем 500 выборок разного размера (15, 30 и 50 элементов) из ГС с ненормальным скошенным Гамма-распределением** 
(т.е. протестируем ЦПТ). Для большей уверенности не только посмотрим на графики, но и проверим гипотезу о том,
что распределение выборочных средних **нормальное** с помощью теста Шапиро-Уилка.


```python
import numpy as np
import seaborn as sns

sample_sizes = [15,30,50]
sample_means = {15:[], 30:[], 50: []}

for n in sample_sizes:
    for _ in range(500):
        sample = np.random.gamma(2., 2., n) # Generating samples from Gamma distribution
        sample_mean = np.mean(sample)
        sample_means[n].append(sample_mean)

df = pd.DataFrame(sample_means)

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,3))
ax1 = sns.histplot(data=df, x=15, kde=True, ax=ax1)
ax2 = sns.histplot(data=df, x=30, kde=True, ax=ax2)
ax3 = sns.histplot(data=df, x=50, kde=True, ax=ax3)
plt.show();
```

Как можно заметить, распределение во всех трех случаях похоже на нормальное.
Проверим это тестом Шапиро-Уилка:

```python
from scipy.stats import shapiro

p_values = {15:[], 30:[], 50: []}
for col in df.columns:
    p_values[col] = shapiro(df[col]).pvalue
p_values
```

Ого! Видим, что в случае, когда мы брали выборки **всего по 15 элементов**,
распределение выборочных средних оказалось **нормальным**,
а в случае, когда выборки большего размера - **ненормальным**.
Конечно, это просто совпадение, и если мы прогоним код еще раз,
результаты могут измениться... но суть остается прежней:
**соблюдение условия n > 30 для t-теста НЕ ТРЕБУЕТСЯ,
но в случае с реальными данными в компании желательно проверять, при каком размере выборки
распределение выборочных средних будет нормальным, т.к. это позволит быть уверенным,
что t-статистика имеет распределение Стьюдента, а значит - результаты t-теста точны**.


#### 2.6.3 Логика теста Шапиро-Уилка

**Гипотезы:**
- $H_0$: выборка извлечена из ГС, в которой признак распределен нормально
- $H_1$: выборка извлечена из ГС, в которой признак распределен не нормально

**Идея теста:**

У нас есть выборка из n элементов (например, из 3), полученная из нормально распределенной ГС.
Мы берем элементы выборки и ранжируем их от наименьшего к наибольшему.
Затем повторяем то же самое много раз и получаем три выборки (т.к. в нашем примере мы многократно берем
из выборки **по 3 элемента**):
- В первой выборке у нас наименьшие числа из трех (числа с рангом 1)
- Во второй выборке у нас средние числа из трех (числа с рангом 2)
- В третьей выборке у нас наибольшие числа из трех (числа с рангом 3)

Если построить гистограмму частот по каждой из полученных выборок,
мы увидим, что эти гистограммы представляют из себя **3 нормальных распределения**,
где самое левое - это первая выборка, центральное - вторая, а самое правое - третья выборка.

Если мы берем **новую выборку из трех элементов**, 
и самый **меньший** ее элемент оказывается близко к **"пику" самого левого распределения**,
**центральный** - близко к **пику центрального распределения**, 
а самый **большой** близко к пику **правого распределения** - эта выборка получена из **нормального распределения**.

Пример с визуализацией в коде:


```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt



def get_gaussian_density(dataset):
    ''' Гауссово распределение, полученное на основе выборки

        :param dataset: массив значений
    '''

    density = ss.gaussian_kde(dataset)
    density.covariance_factor = lambda : 0.25
    density._compute_covariance()

    min_x, max_x = np.percentile(dataset, [0, 100])

    return np.linspace(min_x, max_x, 200), density


mu, sigma = 0, 1 # Среднее и стандартное отклонение
rnd = np.random.normal(mu, sigma, 10000)  # Генеральная совокупность


# Случайно выбираем по 3 пробы, сортируем их
random_samples = (
  sorted(np.random.choice(rnd, 3))
  for i in range(10000)
)

# Zip функция - без комментариев (функциональное программирование, есть на Вики)
# на выходе получаем массивы минимальных, средних и максимальных значений
rnd1, rnd2, rnd3 = zip(*random_samples)


# Гистограмма частот распределения случайной величины (опционально)
plt.figure(figsize=(4,3))
counts_1, x_range_1, ignored = plt.hist(rnd1, 30, density=True, histtype='barstacked')
counts_2, x_range_2, ignored = plt.hist(rnd2, 30, density=True, histtype='barstacked')
counts_3, x_range_3, ignored = plt.hist(rnd3, 30, density=True, histtype='barstacked')


# Гауссова оценка плотности ядра (приближение к гладкой функции)
x1, linear_1 = get_gaussian_density(rnd1)
x2, linear_2 = get_gaussian_density(rnd2)
x3, linear_3 = get_gaussian_density(rnd3)

plt.plot(x1, linear_1(x1))
plt.plot(x2, linear_2(x2))
plt.plot(x3, linear_3(x3))


plt.show();
```

### **2.7 Непараметрические методы - продолжение**

#### 2.7.1 Проблема критерия Шапиро-Уилка

Представим, что есть какая-то ГС с неизвестным распределением.
Мы берем из этой ГС **очень большую выборку (например, n=5000)** и хотим проверить гипотезу о том, 
что **выборка получена из нормального распределения**, с помощью критерия Шапиро-Уилка.

Логично, что **чем больше выборка - тем больше она "похожа" на ГС**, и в нашем случае,
если мы построим **гистограмму частот** по выборке, мы увидим **почти идеальное нормальное распределение**.
Несмотря на "идеальную" гистограмму, мы на всякий случай хотим провести критерий Шапиро-Уилка,
чтобы **убедиться**, что выборка получена из нормального распределения.
Однако, применив критерий Шапиро-Уилка, **мы можем получить p-value < 0.05** (т.е. вынуждены будем отклонить $H_0$)!

Это позволяет сделать один очень важный вывод:
**"идеального"** нормального распределения, по сути, **не существует**, 
т.е. даже если **гистограмма выглядит идеально - это не значит, 
что в выборке нет никаких отклонений от нормального распределения**.
При этом, чем **больше выборка** - тем **сильнее** в ней проявляются эти отклонения от нормального распределения,
и тем **легче критерию Шапиро-Уилка "отловить" эти отклонения и выдать p-value < 0.05** (т.е. чем больше выборка - тем выше мощность теста).

Поэтому, помимо проведения теста Шапиро-Уилка, стоит обращать внимание **не только на p-value**,
но и на **"степень" отклонения эмпирического распределения от нормального**.

Для того, чтобы определить "степень" отклонения эмпирического распределения от нормального,
можно **рассчитать коффициент корреляции Пирсона между эмпирическими и теоретическими квантилями
распределения на графике QQ-plot**. Если корреляция между реальными данными и "идеальными" **больше 0.95 - 
можно считать, что выборка получена из нормального распределения**, даже если тест Шапиро-Уилка выдал p-value < 0.05
и вынудил отклонить нулевую гипотезу.

<img src="images/2.7.1 qqplot and histogram.png" width="500" />

На картинке выше как раз показан описанный случай - **выборка большая**, поэтому тест Шапиро-Уилка имеет **высокую мощность**,
а значит, "отлавливает" даже **очень небольшие отклонения от нормального распределения**. 
В данном случае он "отловил" отклонения **в хвостах распределения** (на графике QQ plot это значения, отклоняющиеся от прямой на краях)
и позволил **отклонить нулевую гипотезу**, несмотря на то, что **корреляция между реальными и идеальными квантилями равна 0.9994!**

**Резюме:**
- Выборку лучше всегда проверять на нормальность распределения тестом Шапиро-Уилка/тестом Колмогорова-Смирнова/корреляцей квантилей на графике QQ plot,
потому что нормальность распределения - это **идеальный вариант**, при котором мы можем вообще не думать об объеме выборки/выборок
- Если выборка получена **не из нормального распределения** - то мы по-прежнему **можем использовать t-тест**,
но в этом случае стоит **подобрать такой размер выборки** (на исторических данных, например), 
чтобы распределение **выборочного среднего было как можно ближе к нормальному**, поскольку именно нормальность распределения выборочного среднего позволяет
принимать решение на основе результатов теста, будучи точно уверенным, что **t-статистика имеет распределение Стьюдента**
- Если оба этих варианта **не подходят** или в выборках есть **выбросы, от которых мы не можем просто избавиться** (т.к. это повлечет за собой
удаление важной для теста информации) или же выборки, которые мы хотим сравнить, имеют совсем **разный размер** - 
**используем непараметрику (U-тест Манна-Уитни как замену t-тесту)**

#### 2.7.2 U-тест Манна-Уитни

Логика U-критерия заключается в том, что вместо сравнения средних значений в двух выборках критерий сравнивает сумму рангов (не медианы, как многие думают). 

**Логика проведения теста:**

Пусть у нас есть две выборки по 6 элементов: A (3,4,2,6,2,5) и B (9,7,5,10,6,8).
- Сначала выписываем **все элементы двух выборок**, упорядочив по возрастанию (2,2,3,4,5,5,6,6,7,8,9,10), 
и каждому элементу проставляем свой **ранг (порядковый номер)**.
Если элемент повторяется (как в данном примере элементы 2,5,6) - проставляем каждому из этих "соседних" элементов ранг, 
равный среднему арифметическому соседних рангов. 
То есть полученная последовательность элементов двух выборок будет иметь такие ранги: (1.5, 1.5, 3, 4, 5.5, 5.5, 7.5, 7.5, 9, 10, 11, 12)
- Затем **для каждой выборки считаем сумму рангов в ней**:
    
    $R_1 = 3+4+1.5+7.5+5.5+1.5 = 23$
    
    $R_2 = 11+9+5.5+12+7.5+10 = 55$
    
- Затем **для каждой из выборок считаем U-статистику**:
    
    $U_1 = R_1 - \frac{n_1(n_1+1)}{2} = 23 - \frac{6(7)}{2} = 2$
    
    $U_2 = R_2 - \frac{n_2(n_2+1)}{2} = 55 - \frac{6(7)}{2} = 34$
- Выбираем **меньшую** из полученных статистик: $U = U_1 = 2$
- Смотрим **критическое значение для U-статистики** и сравниваем полученную статистику с ним,
либо исходим из того, что полученная U-статистика имеет **нормальное распределение** с параметрами 
$\mu = \frac{n_1 \cdot n_2}{2}$ и $\sigma = \sqrt{\frac{n_1 \cdot n_2 \cdot (n_1 + n_2 + 1)}{12}}$
после чего **делаем вывод о гипотезе**.

*Разумно применять, если распределения сравниваемых выборок не нормальны, но **имеют одинаковое направление асимметрии**.

Пример в коде:


```python
from scipy import stats

a = [3,4,2,6,2,5]
b = [9,7,5,10,6,8]

stat, p = stats.mannwhitneyu(a,b)
print(f'U-statistic = {stat}, p-value = {p}')
```

#### 2.7.3 Тест Краскела-Уоллиса

Этот тест является **непараметрическим аналогом дисперсионному анализу (ANOVA)** и позволяет проверять **гипотезу о взаимосвязи между группами**,
если не соблюдаются оба требования для ANOVA:
- Гомогенность дисперсий во всех группах
- Нормальность распределений в группах

По аналогии с t-тестом, если в ANOVA группы получены из ГС с **нормальным распределением** и имеют **равные дисперсии** - мы
можем быть уверены, что **полученная F-статистика имеет F-распределение (Фишера)**.

**Логика теста Краскела-Уоллиса:**
если тест Манна-Уитни сравнивает **средние ранги двух выборок**, и **разность средних рангов двух групп описывается нормальным распределением**,
то статистикой критерия Краскела-Уоллиса является **дисперсия сумм рангов сравниваемых групп**, которая при верной нулевой гипотезе
имеет **распределение хи-квадрат**.

**Формула H-статистики:**

$$H = \frac{12}{N(N+1)}\sum_{i=1}^k \frac{R_i^2}{n_i}-3(N+1), \space N=\sum_{i=1}^k n_i$$

где k - количество групп (выборок)

Как и U-тест Манна-Уитни, тест Краскела-Уоллиса имеет **бОльшую мощность по сравнению с ANOVA**, 
если группы получены из ненормальных распределений и имеют разные дисперсии (слева группы имеют нормальное распределение, справа - хи-квадрат):

<img src="images/2.7.3 anova vs k-y.png" width="500" />

Пример критерия в коде:


```python
from scipy import stats

a = [3,4,2,6,2,5]
b = [9,7,5,10,6,8]
c = [6,1,5,2,10,11]

stat, p = stats.kruskal(a,b,c)
print(f'H-statistic = {stat}, p-value = {p}')
```

```python
# Task 1.
# Начнем с простого и вспомним, как применять логистическую регрессию. 
# Напишите функцию get_coefficients, которая получает на вход dataframe 
# с двумя переменными x ( фактор с произвольным числом градаций) и y ( фактор с двумя градациями).
# Функция строит логистическую модель, где y — зависимая переменная, а x — независимая, и возвращает вектор со значением экспоненты коэффициентов модели.

import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as sf

df = pd.read_csv("https://stepik.org/media/attachments/course/524/test_data_01.csv")

def get_coefficients(df):
    df.y = [0 if i==min(df.y) else 1 for i in df.y] # recoding y from (1,2) to (0,1)
    logreg = sf.glm(formula='y ~ C(x)', data=df, family=sm.families.Binomial()).fit()
    return np.exp(logreg.params)

get_coefficients(df)
```

```python
# Task 2.
# Если в нашей модели есть количественные предикторы, то в интерцепте мы будем иметь значение, 
# соответствующее базовому уровню категориальных предикторов и нулевому уровню количественных. 
# Это не всегда осмысленно. Например, нам не интересен прогноз для людей нулевого возраста или роста. 
# В таких ситуациях количественную переменную имеет смысл предварительно центрировать так, чтобы ноль являлся средним значением переменной. 
# Самый простой способ центрировать переменную — отнять от каждого наблюдения среднее значение всех наблюдений: x_centered_i = x_i - x_mean.

# В этом задании вашей задачей будет написать функцию centered, которая получает на вход датафрейм и имена переменных, 
# которые необходимо центрировать так, как это описано выше. 
# Функция должна возвращать этот же датафрейм, только с центрированными указанными переменными.

import pandas as pd

df = pd.read_csv('https://stepik.org/media/attachments/course/524/cen_data.csv')
var_names = ['X4','X2','X1']
    
def centered(df, var_names):
    for col in df[var_names]:
        df[col] = [i - df[col].mean() for i in df[col]]
        return df
        
centered(df, var_names)
    
```

```python
# Task 3.
# Представьте, что мы работаем в аэропорту в службе безопасности и сканируем багаж пассажиров. 
# В нашем распоряжении есть информация о результатах проверки багажа за предыдущие месяцы. 
# Про каждую вещь мы знаем:
# - являлся ли багаж запрещенным - is_prohibited (No - разрешенный, Yes - запрещенный) 
# - его массу (кг) - weight
# - длину (см) - length
# - ширину (см) - width
# - тип багажа (сумка или чемодан) - type.

# Напишите функцию get_features, которая получает на вход набор данных о багаже. 
# Строит логистическую регрессию, где зависимая переменная - "являлся ли багаж запрещенным", а предикторы - остальные переменные, 
# и возвращает вектор с названиями статистически значимых переменных (p < 0.05) (в модели без взаимодействия). 
# Если в данных нет значимых предикторов, функция возвращает строку с сообщением  "Prediction makes no sense".

import re
import pandas as pd
import numpy as np
import statsmodels.formula.api as sf
import statsmodels.api as sm

df = pd.read_csv('https://stepik.org/media/attachments/course/524/test_luggage_1.csv')
# df.is_prohibited = [0 if i == 'No' else 1 for i in df.is_prohibited] # recoding y from ('No','Yes') to (0,1)

def get_features(df):
    logreg = sf.glm(formula='is_prohibited ~ weight + length + width + C(type)', data=df, family=sm.families.Binomial()).fit()
    pre_result = [i if logreg.summary2().tables[1]['P>|z|'][i] < 0.05 else '' for i in logreg.summary2().tables[1].index]
    result = [re.findall(name, r) for r in pre_result for name in ['weight', 'length', 'width', 'type']]
    result = list(np.array([r for r in result if r]).ravel())
    return result if len(result) else 'Prediction makes no sense'
    
get_features(df)
```

```python
# Task 4.
# Интересной особенностью логистической регрессии является тот факт, что ее предсказания — это не конкретный класс, 
# к которому мы отнесем новое наблюдение, а вероятность отнесения к каждому из классов! 

# В результате, построив регрессионную модель, мы можем сделать вероятностное предсказание для каждого нового наблюдения. 
# Иногда при решении практических задач бывает важным обратить внимание на те объекты, 
# которые получили максимальное значение вероятности принадлежности к одному из классов.

# Продолжим нашу работу в службе безопасности! Разобравшись с тем, какие предикторы могут помогать нам предсказывать запрещенный багаж, 
# давайте применим наши знания для повышения безопасности в аэропорту. 
# Обучим наш алгоритм различать запрещенный и разрешенный багаж на уже имеющихся данных и применим его для сканирования нового багажа!

# Напишите функцию, которая принимает на вход два набора данных. 
# Первый dataframe, как и в предыдущей задаче, содержит информацию об уже осмотренном багаже (запрещенный или нет, вес, длина, ширина, тип сумки). 
# Второй набор данных — это информация о новом багаже, который сканируется прямо сейчас. 
# В данных также есть информация: вес, длина, ширина, тип сумки и имя пассажира (смотри описание переменных в примере). 

# Используя первый набор данных, обучите регрессионную модель различать запрещенный и разрешенный багаж. 
# При помощи полученной модели для каждого наблюдения в новых данных предскажите вероятность того, что багаж является запрещенным. 
# Пассажиров, чей багаж получил максимальное значение вероятности, мы попросим пройти дополнительную проверку. 

# Итого, ваша функция принимает два набора данных и возвращает имя пассажира с наиболее подозрительным багажом. 
# Если несколько пассажиров получили максимальное значение вероятности, то верните вектор с несколькими именами. 

# В этой задаче для предсказания будем использовать все предикторы, даже если некоторые из них оказались незначимыми. 
# Для предсказания стройте модель без взаимодействия предикторов.

import pandas as pd
import statsmodels.formula.api as sf
import statsmodels.api as sm

train_df = pd.read_csv('https://stepik.org/media/attachments/course/524/test_data_passangers.csv')
test_df = pd.read_csv('https://stepik.org/media/attachments/course/524/predict_passangers.csv')
train_df['is_prohibited'] = [1 if i=='Yes' else 0 for i in train_df['is_prohibited']] # recoding Y from (yes/no) to (1/0)

def most_suspicious(test_df, train_df):
    logreg = sf.glm(formula='is_prohibited ~ weight + length + width + C(type)', data=train_df, family=sm.families.Binomial()).fit() # fitting with train data
    y_pred = pd.DataFrame({'p': logreg.predict(test_df[['weight','length','width','type']])}) # prediction of probabilities

    test_df = test_df.merge(y_pred, how='left', left_index=True, right_index=True) # joining predicted probabilities with test data
    
    names = []
    for row in test_df.iterrows():
        if row[1][5] == max(test_df['p']):
            names.append(row[1][4])

    return names if len(names) > 1 else names[0]
    
most_suspicious(test_df, train_df)
    
```

```python
# Task 5.
# Напишите функцию normality_test, которая получает на вход dataframe с произвольным количеством переменных 
# разных типов (количественные, строки, факторы) и проверяет нормальность распределения количественных переменных. 
# Функция должна возвращать вектор значений p-уровней значимости теста shapiro.test для каждой количественной переменной.

from scipy import stats
import pandas as pd

df = pd.read_csv('https://stepik.org/media/attachments/course/524/test.csv')

def norm_test(df):
    numeric_df = df.select_dtypes(include=np.number) # df[df.dtypes[df.dtypes == "float64"].index] - analog
    p_values = {}
    for col in numeric_df.columns:
        p_values[col] = (stats.shapiro(numeric_df[col]).pvalue) # {col: stats.shapiro(df[col]).pvalue for col in numeric_df.columns} - analog
    return p_values

norm_test(df)
```

```python
# Task 6.
# Напишите функцию smart_anova, которая получает на вход dataframe с двумя переменными x и y. 
# Переменная x — это количественная переменная, переменная y - фактор, разбивает наблюдения на три группы.

# Если распределения во всех группах значимо не отличаются от нормального, а дисперсии в группах гомогенны, 
# функция должна сравнить три группы при помощи дисперсионного анализа и вернуть именованный вектор со значением p-value, имя элемента — "ANOVA".

# Если хотя бы в одной группе распределение значимо отличается от нормального или дисперсии негомогенны, 
# функция сравнивает группы при помощи критерия Краскела — Уоллиса и возвращает именованный вектор со значением p-value, имя вектора  — "KW".

# Распределение будем считать значимо отклонившимся от нормального, если в тесте shapiro.test() p < 0.05.

# Дисперсии будем считать не гомогенными, если в тесте bartlett.test() p < 0.05.

from scipy.stats import shapiro, bartlett, kruskal, f_oneway
import pandas as pd
import numpy as np

df = pd.read_csv('https://stepik.org/media/attachments/course/524/s_anova_test.csv')

def smart_anova(df):
    if all(x >= 0.05 for x in [shapiro(df[df['y'] == i]['x']).pvalue for i in df.y.unique()]) & all(x >= 0.05 for x in [bartlett(df[df['y'] == 'A']['x'], df[df['y'] == 'B']['x'], df[df['y'] == 'C']['x'] ).pvalue]):
        return {'ANOVA': f_oneway(df[df['y'] == 'A']['x'], df[df['y'] == 'B']['x'], df[df['y'] == 'C']['x']).pvalue}
    else:
        return {'KW': kruskal(df[df['y'] == 'A']['x'], df[df['y'] == 'B']['x'], df[df['y'] == 'C']['x']).pvalue}
    
smart_anova(df)
```

```python
# Task 7.
# Напишите функцию normality_by, которая принимает на вход dataframe c тремя переменными. 
# Первая переменная количественная, вторая и третья имеют две градации и разбивают наши наблюдения на группы. 
# Функция должна проверять распределение на нормальность в каждой получившейся группе и возвращать dataframe 
# с результатами применения теста shapiro.test (формат вывода смотри ниже).

# Рассмотрим пример работы функции. В данных mtcars есть две группирующие переменные am и vs, 
# и обе переменные имеют две градации 0 и 1. В результате получаем 4 группы:

# am vs
# 1 1
# 1 0
# 0 1
# 0 0

# Теперь для переменной mpg в каждой из четырех групп можно проверить распределение на нормальность при помощи функции shapiro.test
# и вывести получившийся p-value.

from scipy.stats import shapiro
import pandas as pd
import numpy as np

df = pd.read_csv('https://stepik.org/media/attachments/course/524/test_for_norm.csv')
def normality_by(df):
    df_pval = df.groupby(df.columns[1:].to_list())['x'].apply(lambda group: shapiro(group).pvalue).reset_index()
    df_pval.rename(columns={'x': 'p_value'}, inplace=True)
    return df_pval

normality_by(df)
```

## **3. Кластерный анализ и метод главных компонент**

Эти два метода отличаются от классических статистических тестов и используются для решения других задач:
- **Кластерный анализ** позволяет определить, есть ли в наших данных разделение на группы, 
или разделить наши наблюдения на k групп (кластеров)
- **Метод главных компонент (Principal Component Analysis, PCA)** используется для того,
чтобы "объединять" переменные, которые могут быть **взаимосвязаны друг с другом** (
например, **знание статистики** часто могут быть связаны со **знанием Python**,
и PCA позволяет понять, **можно ли объединить эти предикторы в один, тем самым уменьшив общее число предикторов**)

### **3.1 Кластерный анализ методом k-средних**

Кластерный анализ относится к моделям, **обучающихся без учителя**, потому что изначально у нас нет понимания,
есть ли какая-то разбивка на кластеры в данных или нет, а если есть - то сколько этих кластеров.
Один из примеров применения кластерного анализа - это **сегментация пользователей**.
Другой пример - мы открыли новый вид животного и хотим показать, что он может быть разделен на разные подвиды.

**Методы кластеризации:**
- k-means (k-средних)
- иерархическая кластеризация
- другие

#### 3.1.1 Идея метода k-means
Пусть у нас есть наблюдения:

<img src="images/3.1.1 k-means initial.png" width="200"/>

**Алгоритм очень простой:**

1. Сначала мы определяем, **на какое количество кластеров мы хотим разделить наблюдения** (т.е. задаем число **k**)
2. **Случайным образом** алгоритм выбираее начальные позиции **центроидов** на плоскости.

    **Центроид** - это что-то вроде центра тяжести, который находится "в середине" облака точек и имеет координаты, 
    равные **средним значениям наблюдений в облаке по оси X и по оси Y**, т.е. является **геометрическим центром**.
    
    Пусть мы решили разбить данные на k=5 кластеров, после чего алгоритм случайно "выбрасывает" на плоскость 5 будущих центроидов
    и разделяет все наши наблюдения на 5 зон. В каждую зону попадают только те точки, расстояние от которых ближе всего именно до "центроида" этой зоны.
    Другими словами, зеленые точки ближе всего именно к зеленому "центроиду", и аналогично со всеми остальными цветами.
    
    <img src="images/3.1.1 k-means 1.png" width="300"/>
    
3. Алгоритм видит, что **центроиды не находятся в геометрических центрах своих наблюдений**, поэтому двигает их в него, что меняет и форму "зоны".
    Получаем ситуацию, когда центроиды находятся сильно ближе к геометрическим центрам, а "зоны" на плоскости изменились,
    что приводит к тому, что некоторые точки оказались не в своих изначально очерченных зонах (например, часть зеленых точек оказалась в черной зоне,
    часть желтых - в красной и т.д.).

    <img src="images/3.1.1 k-means 2.png" width="300"/>
    
4. Алгоритм видит, что **часть точек нужно "перекрасить"** в новый цвет (т.е. **заново определить,
    к какому центроиду они ближе всего**), потому что они теперь в новой зоне, и перекрашивает их.

    <img src="images/3.1.1 k-means 3.png" width="300"/>
    
5. После "перекрашивания" точек (переопределения, к какому центроиду они ближе всего) алгоритм видит,
    что и **позиции центроидов теперь не совсем в геометрических центрах своих точек**,
    и двигает их к своим новым геометрическим центрам.
    
    <img src="images/3.1.1 k-means 4.png" width="300"/>
    
6. Теперь алгоритм будет **повторять п.4 и п.5** (т.е. двигать центроиды, затем переопределять, к какому из центроидов ближе точки, 
    затем снова двигать центроиды, затем снова "перекрашивать" точки и т.д.) до тех пор, 
    **пока после очередного изменения позиции центроидов не останется ни одной точки, которую нужно "перекрасить"**.
    
    <img src="images/3.1.1 k-means 5.png" width="300"/>
    
Результатом работы алгоритма является то, что **для каждого наблюдения мы можем точно сказать,
к какому кластеру оно относится**.
    
*Для лучшего понимания можно поиграть с [этой визуализацией](https://www.naftaliharris.com/blog/visualizing-k-means-clustering/)

*Во многом **то, насколько "правильно" алгоритм разделяет наблюдения по кластерам, зависит от того,
куда изначально случайным образом попали центроиды**. Более того, иногда мы и сами можем не знать,
какая разбивка "правильная", а какая - нет. 

**Пример явно неправильной разбивки на кластеры:**

<img src="images/3.1.1 k-means bad example.png" width="300"/>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Создаем произвольные 200 точек в диапазоне [-10, 10]
X = -10 + 20 * np.random.rand(200, 2)

# Разбиваем точки на 5 кластеров
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)

# Получаем координаты центров кластеров
centers = kmeans.cluster_centers_

# Получаем метки кластеров для каждой точки
labels = kmeans.predict(X)

# Строим scatter plot, в котором каждый кластер отображается своим цветом
# plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], marker='4', s=200, c='r')
plt.show();
```

В конце предыдущего урока был приведен пример с явно неправильной кластеризацией:

**Всего у нас есть две основных проблемы:**
- Алгоритм **k-means может ошибаться даже в случаях, когда кластеры из точек ярко заметны на плоскости!**
Происходит это в том числе из-за того, что изначально центроиды "раскидываются" по плоскости случайно,
и некоторые из них оказываются слишком близко друг к другу и начинают "делить" между собой одно облако точек.
- На картинке выше хорошо понятно, что алгоритм сошелся неоптимально. 
Но часто мы и сами не можем ответить на вопрос **что значит "правильно"/"оптимально"?**


**Решить эти проблемы можно большим количеством способов, но вот основные два:**
1. **Вместо случайного** первоначального разбрасывания центроидов по плоскости мы можем расположить их так,
чтобы они были **наиболее далеко друг от друга**. Часто (не всегда!) это позволит снизить вероятность того,
что **два центроида "склеются" и будут делить между собой одно облако точек**.
2. Мы можем **проверить, насколько оптимально сошелся алгоритм, прогнав его много раз (например, 100)**.
Если центроиды каждый раз "разбрасываются" по плоскости **случайно, но в большинстве случаев в итоге попадают в схожие места** - 
значит, скорее всего, **это и есть самый оптимальный вариант**, а в данных явно есть некая структура.
Если же центроиды каждый раз в итоге оказываются в разных местах - значит, в данных нет явно выраженной структуры.
Другой вариант - можно попробовать **изменить k**, ибо, возможно, центроидов либо **не хватает для того**,
чтобы "охватить" все облака точек, либо их **слишком много**, из-за чего они "склеиваются".


### **3.3 Как определить оптимальное число кластеров?**

Теперь рассмотрим, **как понять, что данные можно разбить на кластеры**,
и если можно - то **как определить оптимальное количество кластеров?**

Есть множество разных способов ее решения, но самый простой - **методом локтя (с помощью общей внутрикластерной суммы квадратов)**.

*Еще более простой метод - просто **построить диаграмму рассеивания (scatter plot) и визуально оценить наблюдения**,
но этот способ работает **только если у нас один количественный предиктор, а не несколько**
(потому что если их несколько, то придется строить диаграмму для каждой пары предиктор-ЗП,
и диаграмма каждый раз может быть совершенно разная).

**Определения:**
- **Внутрикластерная сумма квадратов (within-cluster sum of squares, WSS)** - это сумма квадратов отклонений каждого наблюдения в кластере от центроида кластера
- **Общая внутрикластерная сумма квадратов (total within-cluster sum of squares, TWSS)** - это сумма всех внутрикластерных сумм квадратов 
(если мы выделяем 3 кластера, то это сумма трех WSS)

**Идея метода локтя:**

Алгоритм k-means сначала разбивает данные на 2 кластера, потом на 3, потом на 4 и т.д., и **каждый раз считает TWSS**.
Чем **меньше расстояния от наблюдений в кластере до центроида этого кластера - тем меньше будет TWSS**,
но при **определенном k** мы увидим, что **с дальнейшим увеличением** кол-ва кластеров k показатель **TWSS не снижается** (или снижается, но очень плавно).
Этот **"перелом"** как раз находится в районе **самого оптимального k**.

- Другими словами, если добавление **еще одного кластера** в данные **значительно понижает общую сумму квадратов TWSS** - это значит,
    что **в добавлении этого кластера есть смысл**. Пример:
    
    <img src="images/3.3 optimal k.png" width="500"/>
    
    В данных на картинке выше если мы выделяем в данных только **два кластера** - **один из центроидов окажется в центре одного облака точек,
    а второй окажется межеду двумя облаками точек, т.е. во втором кластере показатель WSS будет большим,
    что сделает большим и TWSS**. Добавление **третьего** центроида сбалансирует ситуацию и позволит **значительно снизить TWSS**.
    Добавление последующих центроидов будет просто "делить" уже полученные кластеры на более маленькие кластеры,
    что **не окажет значительного влияния на TWSS**.

- Если же мы видим, что **добавление каждого нового кластера понижает TWSS незначительно** (т.е. кривая зависимости TWSS от k достаточно плавная) -
    значит, **в данных невозможно выделить определенное количество кластеров**. Т.е. в этом случае мы можем плавно дойти до ситуации, когда TWSS = 0,
    а количество кластеров k равно количеству наблюдений. Пример:

    <img src="images/3.3 no optimal k.png" width="500"/>

**Важные уточнения:**
- В Python для того, чтобы не определять "на глазок" точку оптимального k, есть библиотека специально для её вычисления - [**kneed**](https://github.com/arvkevi/kneed). 
    
    `from kneed import KneeLocator`
    
- Ещё один вариант - это использовать **коэффициент силуэта (silhouette coefficient)**, 
    который вычисляется с помощью **среднего внутрикластерного расстояния (a)** и **среднего расстояния до ближайшего кластера (b)** по каждому образцу. 
    
    Силуэт вычисляется как **(b - a) / max(a, b)**

    `from sklearn.metrics import silhouette_score`

Пример визуализации метода локтя в коде:


```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

plt.figure(figsize=(4, 4))
sns.set_style('whitegrid')

# Downloading the data 
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
X = df[['sepal length (cm)','petal width (cm)']].values

# Setting MAX number of clusters
total_wcss = []
n_clusters = 10

# Running k-means for each number of clusters and calculating TWSS
for i in range(1, n_clusters): 
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300,  n_init=10, random_state=0).fit(X)
    total_wcss.append(kmeans.inertia_)

# Plotting
plt.plot(range(1, n_clusters), total_wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Total WCSS')
plt.show();
```

### **3.4 Иерархическая кластеризация**

В отличие от метода **k-means**, в котором мы изначально задаем определенное количество кластеров,
**методы иерархической кластеризации** работают по-другому: они сначала рассчитывают расстояния между наблюдениями
(точками на диаграмме рассеивания) и объединяют **близкие друг к другу точки** в кластеры,
затем уже эти кластеры объединяют в бОльшие кластеры,
и так до тех пор, пока все точки не будут объединены в один большой кластер.

**Существует несколько разных методов иерархической кластеризации:**
- Метод одиночной связи (метод ближайшего соседа)
- Метод полной связи (метод дальнего соседа)
- Метод средней связи
- Центроидный метод
- Метод Уорда

#### 3.4.1 Центроидный метод (Centroid Linkage)

<img src="images/3.4.1 k nearest neighbours.png" width="600"/>

**Алгоритм работы метода:**

1. Изначально алгоритм считает, что количество кластеров равно количеству наблюдений N 
    и рассчитывает **расстояния от каждого наблюдения до всех остальных**, формируя таблицу расстояний (на картинке она снизу)
2. Дальше алгоритм объединяет **две самые близкие друг к другу точки в один кластер и считает в этом кластере центроид**
3. Теперь алгоритм **снова рассчитывает расстояния** между всеми наблюдениями,
    но теперь **вместо N наблюдений у нас N-1, т.к. вместо двух ближайших друг к другу точек у нас один центроид**
4. Дальше пункты 2 и 3 **повторяются до тех пор, пока не останется одно "наблюдение" (один кластер)**

На основе полученной дендрограммы мы можем сделать вывод о том, **есть ли вообще смысл кластеризовать данные**, и если есть - то 
**самостоятельно принять решение о том, сколько кластеров мы хотим выделить**.
Например, в этом примере можно выделить **2 кластера: (4,5,7) и (6,8,10,1,2,3,9)**,
а можно, например, **4 кластера: (5), (4,7), (6,8,10) и (1,2,3,9)**.

**Пример данных, в которых нет смысла проводить кластеризацию:**

<img src="images/3.4 clusterization impossible.png" width="600"/>

#### 3.4.2 Метод одиночной связи/ближайшего соседа (Single Linkage/K-Nearest Neighbours)

**Алгоритм работы метода:**

1. Изначально алгоритм считает, что количество кластеров равно количеству наблюдений N 
    и рассчитывает **расстояния от каждого наблюдения до всех остальных**
2. Дальше алгоритм объединяет **две самые близкие друг к другу точки в один кластер**
3. Теперь алгоритм **снова рассчитывает расстояния** между всеми наблюдениями,
    но теперь **вместо N наблюдений у нас N-1, т.к. вместо двух ближайших друг к другу точек алгоритм берет ту,
    расстояние от которой до другой точки/кластера минимально**
4. Дальше пункты 2 и 3 **повторяются до тех пор, пока не останется одно "наблюдение" (один кластер)**

#### 3.4.3 Комбинирование методов

Один из возможнных вариантов - это сначала провести **иерархическую кластеризацию**,
что позволит на основе дендрограммы наглядно определить, сколько кластеров выделить,
а затем с помощью **k-means** объединить наблюдения в кластеры, задав определенное число кластеров.

Другой вариант - **случайно разделить выборку на две части и в каждой части провести иерархическую кластеризацию**.
Если **"структура кластеров" похожа** в двух частях выборки - **есть смысл углубляться и исследовать, на какие кластеры стоит поделить данные**.


```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
import numpy as np

# Для прорисовки графики
%matplotlib inline

# Наши данные
points = np.array([[1,2], [2,1], [2,2], [1,3], [6,7], [6,5], [7,8], [8,8]], dtype=float)

Z = linkage(points, 'ward')
fig = plt.figure(figsize=(6, 3))
dn = dendrogram(Z)

plt.show();
```

### **3.5 Введение в метод анализа главных компонент**

#### 3.5.1 Идея метода главных компонент (PCA)

**Метод анализа главных компонент (Principal Component Analysis, PCA)** - метод снижения размерности данных,
суть которого заключается в "замене" исходных переменных на главные компоненты ("сгруппированные переменные") с потерей определенной информации.
Например, если в исходных данных 10 переменных, с помощью PCA мы можем вместо 10 переменных
использовать всего две новых переменных, потеряв небольшую часть информации о данных.

Разберем подробнее **на примере данных (всего 40 наблюдений) с двумя количественными переменными**:
- **R** - знания языка R (в баллах на курсе)
- **Stat** - знания в области статистики (в баллах на курсе)

<img src="images/3.5.1 data and correlation.png" width="500"/>

Видим, что между R и Stat есть **сильная положительная корреляция**.
Но что позволяет наилучшим способом **описать взаимосвязь между R и Stat**?

Правильно, **регрессионная прямая!**

<img src="images/3.5.1 pc1 and pc2.png" width="500"/>

На картинке выше **синяя прямая - регрессионная прямая**, которая в то же самое время является **первой главной компонентой** или **PC1**.
Красная прямая - это **вторая главная компонента** или **PC2**.
Главные компоненты стоит воспринимать как **новые оси**, например:

- Если мы находимся в исходной системе координат, где ось **X** - это значения переменной R,
    а ось **Y** - значения переменной Stat, то каждое наблюдение мы рассматриваем **с точки зрения проекции на оси X и Y**.
    Соответственно, если мы хотим **предсказать значение Stat, зная значение R** - мы **"поднимаемся" к регрессионной прямой,
    "поворачиваем" на 90 градусов влево и смотрим на значение Y** (т.е. на проекцию значения наблюдения на ось Y).
    
- Если мы находимся в "новой" системе отсчета, где мы воспринимаем **регрессионную прямую/первую главную компоненту как новую "ось X"**,
    то **зная только значение проекции наблюдения на эту ось PC1, мы можем сразу сказать, 
    что наблюдение находится где-то недалеко от PC1 и расположено где-то на прямой, перпендикулярной PC1**.
    
    Более того, оказывается, что в случае, когда **изначально у нас две переменные, PC1 (новая "ось Х") описывает целых ~95% изменчивости наблюдений,
    а оставшиеся ~5% описывает красная прямая, перпендикулярная синей, она же PC2 (новая "ось Y")**.
    
    Соответственно, в этом случае мы можем, например, принять решение **отказаться от PC2, пожертвовав ~5% информации о данных,
    но зато использовать только одну переменную, а не две**.
    
    Сами же главные компоненты можно называть как угодно - например, PC1 можно назвать "Experience in Data Analysis",
    и одной этой переменной мы можем объединить и знания по R, и знания по статистике.
    
**Введение новых осей (главных компонент) позволяет построить график biplot (справа):**

<img src="images/3.5.1 biplot.png" width="500"/>

Можно заметить, что теперь ось X - это PC1, а ось Y - PC2.
Более того, на графике видны векторы R и Stat - т.е. наши исходные переменные.

**Интерпретация biplot:**
- **Угол между R и Stat описывает корреляцию между ними**.
Т.к. угол довольно острый и оба вектора направлены вправо - **корреляция достаточно сильная и положительная**.
Если бы векторы смотрели **в разные стороны - корреляция отрицательная**. Если векторы **перпендикулярны - корреляция равна 0**.
- **Оба вектора R и Stat направлены примерно параллельно оси PC1**.
Это говорит о том, что **изменение значения R или Stat** приводят к **гораздо большему изменению значения оси PC1, чем оси PC2**.
И наоборот, если бы, например **R был вертикально направлен вверх - его изменчивость бы описывалась только осью PC2**.
- **Большое значение по оси PC1 говорит о том, что в этом наблюдении высокое значение и у R, и у Stat (т.к. 
PC1 описывает 95% дисперсии наблюдений). Если же значение по PC1 маленькое - значит, и R, и Stat маленькие.

#### 3.5.2 Зачем все это нужно?

Если у нас всего **две переменные**, то замена их на **одну PC1 нам действительно мало что даст**.
Однако, в ситуации, когда у нас **10 предикторов**, и мы хотим на основе них предсказывать значение ЗП, **это может быть очень полезно!**

Например, если мы строим регрессию и видим, что **некоторые предикторы мультиколлинеарны** (т.е. между ними есть значимая корреляция),
есть смысл в том, чтобы **с помощью PCA объединить одни коррелирующие предикторы в одну главную компоненту, а другие - в другую.**
Это позволит:
- Во-первых, **избавиться от проблемы мультиколлинеарности**
- Во-вторых, **"сгруппировать" предикторы по смыслу**, что позволит формулировать **более интересные выводы**

**Пример:**

Имеются данные о разных регионах Швейцарии с 6 переменными:

<img src="images/3.5.2 data.png" width="400"/>

Проводим анализ главных компонент:

<img src="images/3.5.2 biplot and screeplot.png" width="600"/>

Во-первых, **главных компонент ровно столько же, сколько и исходных переменных**, наша задача - **выбрать, сколько из них мы будем использовать**. 
Помимо этого, как можно заметить, **всего две главные компоненты PC1 и PC2 описывают более 92% дисперсии наблюдений!**
Соответственно, если мы ограничимся **только PC1 и PC2**, мы можем построить, например, biplot,
и делать по нему какие-то интересные выводы, основываясь на векторах (исходных переменных) и наблюдениях, которые также видны на графике 
(например, видно, что Agriculture и Education имеют отрицательную корреляцию, а регион V. De Geneve почему-то сильно отличается от всех остальных).

Если же нас **не устраивают 92% - мы можем использовать 3 главные компоненты,
которые описывают уже более 98% дисперсии.**

```python
# Task 4.
# Напишите функцию get_pca2, которая принимает на вход dataframe с произвольным числом количественных переменных. 
# Функция должна рассчитать, какое минимальное число главных компонент объясняет больше 90% изменчивости в исходных данных 
# и добавлять значения этих компонент в исходный dataframe в виде новых переменных.

import pandas as pd
from sklearn.decomposition import PCA

df = pd.read_csv('https://stepik.org/media/attachments/course/524/pca_test.csv')

def get_pca2(df):
    pca = PCA(n_components=df.shape[1])
    fit = pca.fit_transform(df)
    n_components = sum(np.cumsum(pca.explained_variance_ratio_) < 0.9) + 1
    return pd.concat([df, pd.DataFrame(fit[:, :n_components], columns=[f'PC{i}' for i in range(1, n_components + 1)])], axis=1)

get_pca2(df)
```

```python
# Task 5.
# Как я говорил, метод главных компонент может применяться для борьбы с мультиколлинеарностью в данных 
# (ситуация, когда некоторые переменные очень сильно коррелируют между собой). 
# Однако иногда некоторые переменные не просто сильно взаимосвязаны, но могут представлять линейную комбинацию друг друга. 
# На такие переменные лучше сразу взглянуть повнимательнее и выяснить, откуда они взялись в наших данных.

# Напишите функцию is_multicol, которая получает на вход dataframe произвольного размера с количественными переменными. 
# Функция должна проверять существование строгой мультиколлинеарности, а именно наличие линейной комбинации между предикторами. 
# Линейной комбинацией является ситуация, когда одна переменная может быть выражена через другую переменную при помощи уравнения,
# например, V1 = V2 + 4 или V1 = V2 - 5.

# Функция возвращает имена переменных, между которыми есть линейная зависимость или cобщение "There is no collinearity in the data".

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df1 = pd.read_csv('https://stepik.org/media/attachments/course/524/Norris_1.csv')
df2 = pd.read_csv('https://stepik.org/media/attachments/course/524/Norris_2.csv')
df3 = pd.read_csv('https://stepik.org/media/attachments/course/524/Norris_3.csv')

def is_multicol(df):

    collinear_vars = []
    
    # Перебрать все пары переменных
    for i in range(len(df.columns)):
        for j in range(i+1, len(df.columns)):
            
            # Fitting a linear regression to check if residuals are close to zero
            X = df[df.columns[i]].values.reshape(-1, 1)
            y = df[df.columns[j]].values
            
            # Calculating the residuals (difference between predicted and real values)
            y_pred = LinearRegression().fit(X, y).predict(X)
            residuals = y - y_pred
            
            # If all residuals for each pair of columns are close to zero - there is a linear dependence
            if all(np.isclose(abs(res), 0) for res in residuals):
                collinear_vars.append([df.columns[i], df.columns[j]])
                
    return [var for pair in collinear_vars for var in pair] if collinear_vars else 'There is no collinearity in the data'

is_multicol(df3)

```
